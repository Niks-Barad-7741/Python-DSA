# 🌳 PART 1 — Binary Search Tree (BST)

## What is a Tree?
A Tree is a **hierarchical data structure** where each node has at most 2 children.

```
          [10]          ← Root
         /    \
       [5]    [15]      ← Parent nodes
       / \    /  \
     [3] [7][12] [20]   ← Leaf nodes
```

## BST Rules:
- Left child is **SMALLER** than parent
- Right child is **LARGER** than parent
- Search becomes **O(log n)**!

## Traversals:
| Type | Order | Use |
|------|-------|-----|
| Inorder | Left→Root→Right | Sorted output |
| Preorder | Root→Left→Right | Copy tree |
| Postorder | Left→Right→Root | Delete tree |

## Big O:
| Operation | Average | Worst |
|-----------|---------|-------|
| Search | O(log n) ⚡ | O(n) |
| Insert | O(log n) ⚡ | O(n) |
| Delete | O(log n) ⚡ | O(n) |

---

````python
# ============================================
# Binary Search Tree — Full Implementation
# ============================================

class Node:
    def __init__(self, value):
        self.value = value
        self.left  = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, value):                         # O(log n) avg
        if self.root is None:
            self.root = Node(value)
        else:
            self._insert(self.root, value)

    def _insert(self, node, value):
        if value < node.value:
            if node.left is None: node.left = Node(value)
            else: self._insert(node.left, value)
        else:
            if node.right is None: node.right = Node(value)
            else: self._insert(node.right, value)

    def search(self, value):                         # O(log n) avg
        return self._search(self.root, value)

    def _search(self, node, value):
        if node is None: return False
        if node.value == value: return True
        elif value < node.value: return self._search(node.left, value)
        else: return self._search(node.right, value)

    def inorder(self):                               # Left→Root→Right (sorted!)
        result = []
        self._inorder(self.root, result)
        return result

    def _inorder(self, node, result):
        if node:
            self._inorder(node.left, result)
            result.append(node.value)
            self._inorder(node.right, result)

    def preorder(self):                              # Root→Left→Right
        result = []
        self._preorder(self.root, result)
        return result

    def _preorder(self, node, result):
        if node:
            result.append(node.value)
            self._preorder(node.left, result)
            self._preorder(node.right, result)

    def postorder(self):                             # Left→Right→Root
        result = []
        self._postorder(self.root, result)
        return result

    def _postorder(self, node, result):
        if node:
            self._postorder(node.left, result)
            self._postorder(node.right, result)
            result.append(node.value)

    def height(self):
        return self._height(self.root)

    def _height(self, node):
        if node is None: return -1
        return 1 + max(self._height(node.left), self._height(node.right))

    def display(self):
        print("Inorder (sorted):", self.inorder())
        print("Preorder:        ", self.preorder())
        print("Postorder:       ", self.postorder())
        print("Height:          ", self.height())


# TEST
bst = BinarySearchTree()
for v in [10, 5, 15, 3, 7, 12, 20]:
    bst.insert(v)

print("=== BINARY SEARCH TREE ===")
bst.display()

print("\n=== SEARCH ===")
print(f"Search 7:  {bst.search(7)}")
print(f"Search 99: {bst.search(99)}")

print("\n=== INSERT NEW VALUE ===")
bst.insert(8)
print("After inserting 8:")
print("Inorder:", bst.inorder())
````

# ⚖️ PART 2 — AVL Tree (Self-Balancing)

## What is an AVL Tree?
AVL = Self-Balancing BST. It **automatically rotates** nodes to keep the tree balanced!

```
Normal BST (sorted input):    AVL Tree (same input):
[1]                                 [3]
  \                                /   \
  [2]                            [2]   [4]
    \                            /       \
    [3]                        [1]       [5]
      \
      [4]        Height=4      Height=2  ← O(log n) always!
        \
        [5]
```

## Balance Factor:
```
Balance = Height(left) - Height(right)
Allowed: -1, 0, +1
If ±2 → ROTATE to fix!
```

## 4 Rotation Types:
| Case | Rotation |
|------|----------|
| Left-Left | Right Rotation |
| Right-Right | Left Rotation |
| Left-Right | Left then Right |
| Right-Left | Right then Left |

---

````python
# ============================================
# AVL Tree — Self Balancing
# ============================================

class AVLNode:
    def __init__(self, value):
        self.value  = value
        self.left   = None
        self.right  = None
        self.height = 1

class AVLTree:
    def __init__(self):
        self.root = None

    def _height(self, node):         return node.height if node else 0
    def _balance(self, node):        return self._height(node.left) - self._height(node.right) if node else 0
    def _update_height(self, node):  node.height = 1 + max(self._height(node.left), self._height(node.right))

    def _rotate_right(self, y):
        x = y.left; T2 = x.right
        x.right = y; y.left = T2
        self._update_height(y); self._update_height(x)
        return x

    def _rotate_left(self, x):
        y = x.right; T2 = y.left
        y.left = x; x.right = T2
        self._update_height(x); self._update_height(y)
        return y

    def insert(self, value):
        self.root = self._insert(self.root, value)

    def _insert(self, node, value):
        if node is None: return AVLNode(value)
        if value < node.value:   node.left  = self._insert(node.left, value)
        elif value > node.value: node.right = self._insert(node.right, value)
        else: return node

        self._update_height(node)
        balance = self._balance(node)

        if balance > 1  and value < node.left.value:                             # LL
            print(f"  Right rotation at {node.value}!")
            return self._rotate_right(node)
        if balance < -1 and value > node.right.value:                            # RR
            print(f"  Left rotation at {node.value}!")
            return self._rotate_left(node)
        if balance > 1  and value > node.left.value:                             # LR
            print(f"  Left-Right rotation at {node.value}!")
            node.left = self._rotate_left(node.left)
            return self._rotate_right(node)
        if balance < -1 and value < node.right.value:                            # RL
            print(f"  Right-Left rotation at {node.value}!")
            node.right = self._rotate_right(node.right)
            return self._rotate_left(node)
        return node

    def inorder(self):
        result = []
        self._inorder(self.root, result)
        return result

    def _inorder(self, node, result):
        if node:
            self._inorder(node.left, result)
            result.append(node.value)
            self._inorder(node.right, result)

    def height(self):       return self._height(self.root)
    def root_balance(self): return self._balance(self.root)


# TEST
print("=== AVL TREE — Self Balancing ===")
print("\n--- Inserting sorted data (worst case for BST!) ---")

avl = AVLTree()
for v in [1, 2, 3, 4, 5]:
    print(f"Inserting {v}...")
    avl.insert(v)

print(f"\nInorder:        {avl.inorder()}")
print(f"Tree height:    {avl.height()}")
print(f"Root balance:   {avl.root_balance()}")
print(f"\nBST height would be: 4")
print(f"AVL height is:       {avl.height()}")
print(f"AVL saves: {4 - avl.height()} levels! ")
````

# 🏔️ PART 3 — Heaps

## What is a Heap?
A Heap is a **Complete Binary Tree** with an ordering rule!

```
MAX HEAP → Parent always LARGER than children
        [50]
       /    \
     [30]  [40]
     / \   /
   [10][20][35]

MIN HEAP → Parent always SMALLER than children
        [10]
       /    \
     [20]  [30]
```

## Stored as Array!
```
Parent of i   = (i-1) // 2
Left child    = 2*i + 1
Right child   = 2*i + 2
```

## Big O:
| Operation | Big O |
|-----------|-------|
| Insert | O(log n) ⚡ |
| Get max/min | O(1) ⚡ |
| Remove max/min | O(log n) ⚡ |
| Build heap | O(n) |

---

````python
# ============================================
# Max Heap from Scratch
# ============================================

class MaxHeap:
    def __init__(self):
        self.heap = []

    def _parent(self, i): return (i - 1) // 2
    def _left(self, i):   return 2 * i + 1
    def _right(self, i):  return 2 * i + 2

    def insert(self, value):                     # O(log n)
        self.heap.append(value)
        self._bubble_up(len(self.heap) - 1)

    def _bubble_up(self, i):
        while i > 0:
            parent = self._parent(i)
            if self.heap[i] > self.heap[parent]:
                self.heap[i], self.heap[parent] = self.heap[parent], self.heap[i]
                i = parent
            else:
                break

    def get_max(self):                           # O(1)
        return self.heap[0] if self.heap else "Empty!"

    def remove_max(self):                        # O(log n)
        if not self.heap: return "Empty!"
        if len(self.heap) == 1: return self.heap.pop()
        max_val = self.heap[0]
        self.heap[0] = self.heap.pop()
        self._bubble_down(0)
        return max_val

    def _bubble_down(self, i):
        size = len(self.heap)
        while True:
            largest = i
            l, r = self._left(i), self._right(i)
            if l < size and self.heap[l] > self.heap[largest]: largest = l
            if r < size and self.heap[r] > self.heap[largest]: largest = r
            if largest != i:
                self.heap[i], self.heap[largest] = self.heap[largest], self.heap[i]
                i = largest
            else: break

    def size(self):     return len(self.heap)
    def is_empty(self): return len(self.heap) == 0
    def display(self):  print(f"Heap: {self.heap}")


# TEST Max Heap
print("=== MAX HEAP ===")
mh = MaxHeap()
for v in [10, 30, 20, 50, 40, 35]:
    mh.insert(v)
    print(f"Insert {v:2} → {mh.heap}")

print(f"\nMax value: {mh.get_max()}")
print("\nRemoving max values:")
while not mh.is_empty():
    print(f"  Removed: {mh.remove_max()}")
````

````python
# ============================================
# Python heapq (built-in Min Heap)
# ============================================
import heapq

print("=== PYTHON heapq (MIN HEAP) ===")
heap = []
for v in [30, 10, 50, 20, 40]:
    heapq.heappush(heap, v)
    print(f"Push {v:2} → {heap}")

print(f"\nMin value: {heap[0]}")
print("\nPopping min values:")
while heap:
    print(f"  Popped: {heapq.heappop(heap)}")

print("\n=== MAX HEAP using heapq (negate trick!) ===")
max_heap = []
for v in [30, 10, 50, 20, 40]:
    heapq.heappush(max_heap, -v)
print("Popping max values:")
while max_heap:
    print(f"  Popped: {-heapq.heappop(max_heap)}")
````

````python
# ============================================
# Real World — Hospital Priority Queue
# ============================================
import heapq

print("=== HOSPITAL EMERGENCY ROOM ===")
print("Priority: 1=Critical, 2=Serious, 3=Normal\n")

er = []
patients = [
    (3, "Nikhil  - Headache"),
    (1, "Priya   - Heart Attack"),
    (2, "Sara    - Broken Arm"),
    (1, "Raj     - Stroke"),
    (3, "Divya   - Fever"),
]
for priority, patient in patients:
    heapq.heappush(er, (priority, patient))
    print(f"Admitted: [{priority}] {patient}")

print("\nTreating patients (highest priority first):")
while er:
    priority, patient = heapq.heappop(er)
    label = " CRITICAL" if priority==1 else " SERIOUS" if priority==2 else " NORMAL"
    print(f"  {label} → {patient}")
````

# 🔤 PART 4 — Tries (Prefix Tree)

## What is a Trie?
A Trie stores **strings** where each node is a **character**!

```
Words: cat, car, card, bat

        root
       /    \
      c      b
      |      |
      a      a
     / \     |
    t   r    t
        |
        d

Shared prefix 'ca' stored ONCE! ✅
```

## Big O (m = word length):
| Operation | Big O |
|-----------|-------|
| Insert | O(m) ⚡ |
| Search | O(m) ⚡ |
| Prefix search | O(m) ⚡ |
| Autocomplete | O(m + results) |

---

````python
# ============================================
# Trie — Full Implementation
# ============================================

class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end   = False

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):                      # O(m)
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end = True

    def search(self, word):                      # O(m)
        node = self.root
        for char in word:
            if char not in node.children: return False
            node = node.children[char]
        return node.is_end

    def starts_with(self, prefix):               # O(m)
        node = self.root
        for char in prefix:
            if char not in node.children: return False
            node = node.children[char]
        return True

    def autocomplete(self, prefix):              # O(m + results)
        node = self.root
        for char in prefix:
            if char not in node.children: return []
            node = node.children[char]
        results = []
        self._dfs(node, prefix, results)
        return results

    def _dfs(self, node, current, results):
        if node.is_end: results.append(current)
        for char, child in node.children.items():
            self._dfs(child, current + char, results)

    def delete(self, word):
        def _delete(node, word, depth):
            if depth == len(word):
                if not node.is_end: return False
                node.is_end = False
                return len(node.children) == 0
            char = word[depth]
            if char not in node.children: return False
            if _delete(node.children[char], word, depth+1):
                del node.children[char]
                return len(node.children) == 0
            return False
        _delete(self.root, word, 0)

    def all_words(self):
        results = []
        self._dfs(self.root, "", results)
        return sorted(results)


# TEST Trie
print("=== TRIE ===")
trie = Trie()
words = ["cat", "car", "card", "care", "bat", "ball", "banana"]
for w in words: trie.insert(w)
print(f"Inserted: {words}")
print(f"\nAll words: {trie.all_words()}")

print("\n--- search() ---")
print(f"Search 'cat':   {trie.search('cat')}")
print(f"Search 'ca':    {trie.search('ca')}")
print(f"Search 'cards': {trie.search('cards')}")

print("\n--- starts_with() ---")
print(f"Prefix 'ca':  {trie.starts_with('ca')}")
print(f"Prefix 'xyz': {trie.starts_with('xyz')}")

print("\n--- autocomplete() ---")
print(f"'ca' → {trie.autocomplete('ca')}")
print(f"'ba' → {trie.autocomplete('ba')}")

print("\n--- delete() ---")
trie.delete("card")
print(f"After deleting 'card': {trie.autocomplete('car')}")
````

````python
# ============================================
# Real World — Google Autocomplete
# ============================================
print("=== GOOGLE SEARCH AUTOCOMPLETE ===\n")

search_trie = Trie()
searches = [
    "python", "python tutorial", "python list",
    "python dictionary", "python for beginners",
    "data science", "data structures", "data analysis",
    "machine learning", "machine learning tutorial",
    "deep learning", "deep learning python"
]
for s in searches: search_trie.insert(s)

for query in ["py", "data", "machine", "deep", "xyz"]:
    suggestions = search_trie.autocomplete(query)
    print(f"User types '{query}':")
    if suggestions:
        for s in suggestions: print(f"   {s}")
    else:
        print(f"  No suggestions found!")
    print()
````

# 🕸️ PART 5 — Graphs

## What is a Graph?
A Graph is a collection of **nodes (vertices)** connected by **edges**!

```
Nikhil ──── Priya
  |         /   \
Sara ── Raj    Divya
```

## Types:
| Type | Description |
|------|-------------|
| Undirected | Edges go both ways |
| Directed | Edges go one way |
| Weighted | Edges have cost/distance |
| Cyclic | Has loops |

## Traversals:
| Type | Uses | Order |
|------|------|-------|
| BFS | Queue | Level by level |
| DFS | Stack/Recursion | Deep first |

## Big O (V=vertices, E=edges):
| Operation | Adj List |
|-----------|----------|
| Add vertex | O(1) |
| Add edge | O(1) |
| BFS / DFS | O(V+E) |

---

````python
# ============================================
# Graph — Full Implementation
# ============================================
from collections import deque

class Graph:
    def __init__(self, directed=False):
        self.adjacency_list = {}
        self.directed = directed

    def add_vertex(self, v):
        if v not in self.adjacency_list:
            self.adjacency_list[v] = []

    def add_edge(self, v1, v2):
        self.add_vertex(v1); self.add_vertex(v2)
        self.adjacency_list[v1].append(v2)
        if not self.directed:
            self.adjacency_list[v2].append(v1)

    def remove_edge(self, v1, v2):
        self.adjacency_list[v1].remove(v2)
        if not self.directed:
            self.adjacency_list[v2].remove(v1)

    def has_edge(self, v1, v2):
        return v2 in self.adjacency_list.get(v1, [])

    def neighbors(self, v):
        return self.adjacency_list.get(v, [])

    def bfs(self, start):                        # Level by level! Uses Queue
        visited = set([start])
        queue   = deque([start])
        result  = []
        while queue:
            v = queue.popleft()
            result.append(v)
            for n in self.adjacency_list[v]:
                if n not in visited:
                    visited.add(n)
                    queue.append(n)
        return result

    def dfs(self, start):                        # Deep first! Uses Recursion
        visited = set()
        result  = []
        def _dfs(v):
            visited.add(v); result.append(v)
            for n in self.adjacency_list[v]:
                if n not in visited: _dfs(n)
        _dfs(start)
        return result

    def has_path(self, start, end):
        visited = set()
        def _dfs(v):
            if v == end: return True
            visited.add(v)
            for n in self.adjacency_list[v]:
                if n not in visited and _dfs(n): return True
            return False
        return _dfs(start)

    def display(self):
        print("Graph (Adjacency List):")
        for v, neighbors in self.adjacency_list.items():
            print(f"  {v} → {neighbors}")


# TEST Undirected Graph
print("=== UNDIRECTED GRAPH ===")
g = Graph()
for v1,v2 in [("Nikhil","Priya"),("Nikhil","Sara"),("Priya","Raj"),("Priya","Divya"),("Sara","Raj")]:
    g.add_edge(v1, v2)
g.display()

print(f"\nNeighbors of Nikhil: {g.neighbors('Nikhil')}")
print(f"Has edge Nikhil→Raj: {g.has_edge('Nikhil','Raj')}")
print(f"\nBFS from Nikhil: {g.bfs('Nikhil')}")
print(f"DFS from Nikhil: {g.dfs('Nikhil')}")
print(f"\nPath Nikhil→Divya: {g.has_path('Nikhil','Divya')}")
print(f"Path Nikhil→Karan: {g.has_path('Nikhil','Karan') if 'Karan' in g.adjacency_list else False}")
````

````python
# ============================================
# Directed & Weighted Graph
# ============================================
print("=== DIRECTED GRAPH (Instagram follows) ===")
instagram = Graph(directed=True)
instagram.add_edge("Nikhil", "Priya")
instagram.add_edge("Nikhil", "Sara")
instagram.add_edge("Priya",  "Nikhil")
instagram.add_edge("Sara",   "Raj")
instagram.display()
print(f"\nNikhil follows Priya: {instagram.has_edge('Nikhil','Priya')}")
print(f"Priya follows Sara:   {instagram.has_edge('Priya','Sara')}")

print("\n=== WEIGHTED GRAPH (City distances km) ===")
city_graph = {
    "Ahmedabad": [("Mumbai", 530), ("Delhi", 950)],
    "Mumbai":    [("Ahmedabad", 530), ("Pune", 150)],
    "Delhi":     [("Ahmedabad", 950), ("Jaipur", 280)],
    "Pune":      [("Mumbai", 150)],
    "Jaipur":    [("Delhi", 280)],
}
for city, connections in city_graph.items():
    for neighbor, dist in connections:
        print(f"  {city} ──{dist}km──→ {neighbor}")
````

# 🏆 Trees & Graphs — Complete Summary

| Structure | Best For | Search | Insert |
|-----------|----------|--------|--------|
| BST | Ordered data | O(log n) avg | O(log n) avg |
| AVL Tree | Always balanced | O(log n) always | O(log n) |
| Max Heap | Get largest fast | O(n) | O(log n) |
| Min Heap | Get smallest fast | O(n) | O(log n) |
| Trie | String/prefix search | O(m) | O(m) |
| Graph | Networks/relationships | O(V+E) BFS/DFS | O(1) |

## Real World Uses:
```
BST      → databases, file systems
AVL      → databases needing guaranteed O(log n)
Heap     → priority queues, scheduling, Dijkstra
Trie     → autocomplete, spell checker, IP routing
Graph    → social networks, maps, recommendation systems
```
