````python
# ============================================
# CELL 1 — Binary Search Tree
# ============================================

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None     # left child
        self.right = None    # right child

class BinarySearchTree:
    def __init__(self):
        self.root = None

    # O(log n) avg — Insert
    def insert(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            self._insert(self.root, value)

    def _insert(self, node, value):
        if value < node.value:              # go LEFT!
            if node.left is None:
                node.left = Node(value)
            else:
                self._insert(node.left, value)
        else:                               # go RIGHT!
            if node.right is None:
                node.right = Node(value)
            else:
                self._insert(node.right, value)

    # O(log n) avg — Search
    def search(self, value):
        return self._search(self.root, value)

    def _search(self, node, value):
        if node is None:
            return False                    # not found!
        if node.value == value:
            return True                     # found!
        elif value < node.value:
            return self._search(node.left, value)   # go LEFT!
        else:
            return self._search(node.right, value)  # go RIGHT!

    # Inorder Traversal — gives sorted order!
    def inorder(self):
        result = []
        self._inorder(self.root, result)
        return result

    def _inorder(self, node, result):
        if node:
            self._inorder(node.left, result)   # left first
            result.append(node.value)          # then root
            self._inorder(node.right, result)  # then right

    # Preorder Traversal
    def preorder(self):
        result = []
        self._preorder(self.root, result)
        return result

    def _preorder(self, node, result):
        if node:
            result.append(node.value)           # root first
            self._preorder(node.left, result)   # then left
            self._preorder(node.right, result)  # then right

    # Postorder Traversal
    def postorder(self):
        result = []
        self._postorder(self.root, result)
        return result

    def _postorder(self, node, result):
        if node:
            self._postorder(node.left, result)   # left first
            self._postorder(node.right, result)  # then right
            result.append(node.value)            # root last

    # Height of tree
    def height(self):
        return self._height(self.root)

    def _height(self, node):
        if node is None:
            return -1
        left_h  = self._height(node.left)
        right_h = self._height(node.right)
        return 1 + max(left_h, right_h)

    # Display tree (simple)
    def display(self):
        print("Inorder (sorted):", self.inorder())
        print("Preorder:        ", self.preorder())
        print("Postorder:       ", self.postorder())
        print("Height:          ", self.height())


# ============================================
# TEST BST!
# ============================================
bst = BinarySearchTree()

values = [10, 5, 15, 3, 7, 12, 20]
for v in values:
    bst.insert(v)

print("=== BINARY SEARCH TREE ===")
bst.display()

print("\n=== SEARCH ===")
print(f"Search 7:  {bst.search(7)}")    # True
print(f"Search 99: {bst.search(99)}")   # False

print("\n=== INSERT NEW VALUE ===")
bst.insert(8)
print("After inserting 8:")
print("Inorder:", bst.inorder())        # 8 appears between 7 and 10!
````

````python
# ============================================
# CELL 1 — AVL Tree
# ============================================

class AVLNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.height = 1       # height of this node!

class AVLTree:
    def __init__(self):
        self.root = None

    # Get height of node
    def _height(self, node):
        return node.height if node else 0

    # Get balance factor
    def _balance(self, node):
        return self._height(node.left) - self._height(node.right) if node else 0

    # Update height
    def _update_height(self, node):
        node.height = 1 + max(self._height(node.left),
                               self._height(node.right))

    # RIGHT ROTATION
    def _rotate_right(self, y):
        x  = y.left
        T2 = x.right

        x.right = y        # rotate!
        y.left  = T2

        self._update_height(y)
        self._update_height(x)
        return x           # x is new root!

    # LEFT ROTATION
    def _rotate_left(self, x):
        y  = x.right
        T2 = y.left

        y.left  = x        # rotate!
        x.right = T2

        self._update_height(x)
        self._update_height(y)
        return y           # y is new root!

    # Insert and auto-balance!
    def insert(self, value):
        self.root = self._insert(self.root, value)

    def _insert(self, node, value):
        # Normal BST insert
        if node is None:
            return AVLNode(value)
        if value < node.value:
            node.left = self._insert(node.left, value)
        elif value > node.value:
            node.right = self._insert(node.right, value)
        else:
            return node    # duplicates not allowed!

        # Update height
        self._update_height(node)

        # Check balance
        balance = self._balance(node)

        # LEFT LEFT → Right Rotation
        if balance > 1 and value < node.left.value:
            print(f"  Right rotation at {node.value}!")
            return self._rotate_right(node)

        # RIGHT RIGHT → Left Rotation
        if balance < -1 and value > node.right.value:
            print(f"  Left rotation at {node.value}!")
            return self._rotate_left(node)

        # LEFT RIGHT → Left then Right Rotation
        if balance > 1 and value > node.left.value:
            print(f"  Left-Right rotation at {node.value}!")
            node.left = self._rotate_left(node.left)
            return self._rotate_right(node)

        # RIGHT LEFT → Right then Left Rotation
        if balance < -1 and value < node.right.value:
            print(f"  Right-Left rotation at {node.value}!")
            node.right = self._rotate_right(node.right)
            return self._rotate_left(node)

        return node

    # Inorder traversal
    def inorder(self):
        result = []
        self._inorder(self.root, result)
        return result

    def _inorder(self, node, result):
        if node:
            self._inorder(node.left, result)
            result.append(node.value)
            self._inorder(node.right, result)

    # Height of tree
    def height(self):
        return self._height(self.root)

    # Balance factor of root
    def root_balance(self):
        return self._balance(self.root)


# ============================================
# TEST AVL!
# ============================================
print("=== AVL TREE — Self Balancing ===")
print("\n--- Inserting sorted data (worst case for BST!) ---")

avl = AVLTree()
for v in [1, 2, 3, 4, 5]:
    print(f"Inserting {v}...")
    avl.insert(v)

print(f"\nInorder:        {avl.inorder()}")
print(f"Tree height:    {avl.height()}")        # 2, not 4!
print(f"Root balance:   {avl.root_balance()}")  # -1, 0 or 1!

print("\n--- Compare with normal BST ---")
print("BST height with [1,2,3,4,5] would be: 4")
print(f"AVL height with [1,2,3,4,5] is:       {avl.height()}")
print(f"AVL saves:      {4 - avl.height()} levels! ")

### AVL vs BST — Key difference:
# ```
# BST → fast but can become unbalanced!
# AVL → always balanced → always O(log n)!

# BST Search → O(log n) average, O(n) worst
# AVL Search → O(log n) ALWAYS! ✅
````

````python
# ============================================
# CELL 1 — Max Heap from Scratch
# ============================================

class MaxHeap:
    def __init__(self):
        self.heap = []

    def _parent(self, i):     return (i - 1) // 2
    def _left(self, i):       return 2 * i + 1
    def _right(self, i):      return 2 * i + 2

    # O(log n) — Insert
    def insert(self, value):
        self.heap.append(value)       # add at end
        self._bubble_up(len(self.heap) - 1)

    def _bubble_up(self, i):
        # If child > parent → swap! (bubble up)
        while i > 0:
            parent = self._parent(i)
            if self.heap[i] > self.heap[parent]:
                self.heap[i], self.heap[parent] = \
                self.heap[parent], self.heap[i]
                i = parent
            else:
                break

    # O(1) — Get max (root!)
    def get_max(self):
        if not self.heap:
            return "Heap is empty!"
        return self.heap[0]

    # O(log n) — Remove max
    def remove_max(self):
        if not self.heap:
            return "Heap is empty!"
        if len(self.heap) == 1:
            return self.heap.pop()

        max_val = self.heap[0]
        self.heap[0] = self.heap.pop()  # move last to root
        self._bubble_down(0)
        return max_val

    def _bubble_down(self, i):
        size = len(self.heap)
        while True:
            largest = i
            left    = self._left(i)
            right   = self._right(i)

            if left  < size and self.heap[left]  > self.heap[largest]:
                largest = left
            if right < size and self.heap[right] > self.heap[largest]:
                largest = right

            if largest != i:
                self.heap[i], self.heap[largest] = \
                self.heap[largest], self.heap[i]
                i = largest
            else:
                break

    def size(self):    return len(self.heap)
    def is_empty(self):return len(self.heap) == 0
    def display(self): print(f"Heap array: {self.heap}")


# ============================================
# TEST Max Heap!
# ============================================
print("=== MAX HEAP ===")
mh = MaxHeap()

for v in [10, 30, 20, 50, 40, 35]:
    mh.insert(v)
    print(f"Insert {v:2} → {mh.heap}")

print(f"\nMax value:  {mh.get_max()}")    # 50!
print(f"Size:       {mh.size()}")

print("\nRemoving max values one by one:")
while not mh.is_empty():
    print(f"  Removed: {mh.remove_max()}")
# Prints: 50, 40, 35, 30, 20, 10 (descending!)
````

````python
# ============================================
# CELL 2 — Python heapq (built-in Min Heap!)
# ============================================
import heapq

print("=== PYTHON heapq (MIN HEAP) ===")

heap = []

# Insert — O(log n)
values = [30, 10, 50, 20, 40]
for v in values:
    heapq.heappush(heap, v)
    print(f"Push {v:2} → {heap}")

print(f"\nMin value: {heap[0]}")    # always index 0!

print("\nPopping min values:")
while heap:
    print(f"  Popped: {heapq.heappop(heap)}")
# Prints: 10, 20, 30, 40, 50 (ascending!)

# ============================================
# Max Heap using heapq (negate values trick!)
# ============================================
print("\n=== MAX HEAP using heapq ===")
max_heap = []
for v in [30, 10, 50, 20, 40]:
    heapq.heappush(max_heap, -v)    # negate!

print("Popping max values:")
while max_heap:
    print(f"  Popped: {-heapq.heappop(max_heap)}")  # negate back!
# Prints: 50, 40, 30, 20, 10
````

````python
# ============================================
# CELL 1 — Trie from Scratch
# ============================================

class TrieNode:
    def __init__(self):
        self.children = {}       # char → TrieNode
        self.is_end = False      # marks end of word!

class Trie:
    def __init__(self):
        self.root = TrieNode()

    # O(m) — Insert word
    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end = True       # mark end of word!

    # O(m) — Search exact word
    def search(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                return False     # character missing!
            node = node.children[char]
        return node.is_end       # is it a complete word?

    # O(m) — Check if any word starts with prefix
    def starts_with(self, prefix):
        node = self.root
        for char in prefix:
            if char not in node.children:
                return False
            node = node.children[char]
        return True              # prefix exists!

    # O(m) — Get ALL words with given prefix (autocomplete!)
    def autocomplete(self, prefix):
        node = self.root
        for char in prefix:
            if char not in node.children:
                return []        # no words with this prefix!
            node = node.children[char]

        # Now find all words from this point!
        results = []
        self._dfs(node, prefix, results)
        return results

    def _dfs(self, node, current, results):
        if node.is_end:
            results.append(current)
        for char, child in node.children.items():
            self._dfs(child, current + char, results)

    # O(m) — Delete word
    def delete(self, word):
        def _delete(node, word, depth):
            if depth == len(word):
                if not node.is_end:
                    return False    # word not found!
                node.is_end = False
                return len(node.children) == 0  # delete if no children
            char = word[depth]
            if char not in node.children:
                return False        # word not found!
            should_delete = _delete(node.children[char], word, depth + 1)
            if should_delete:
                del node.children[char]
                return len(node.children) == 0
            return False
        _delete(self.root, word, 0)

    # Get all words in trie
    def all_words(self):
        results = []
        self._dfs(self.root, "", results)
        return sorted(results)


# ============================================
# TEST Trie!
# ============================================
print("=== TRIE ===")
trie = Trie()

words = ["cat", "car", "card", "care", "bat", "ball", "banana"]
for word in words:
    trie.insert(word)
print(f"Inserted: {words}")

print(f"\nAll words: {trie.all_words()}")

print("\n--- search() ---")
print(f"Search 'cat':    {trie.search('cat')}")    # True
print(f"Search 'car':    {trie.search('car')}")    # True
print(f"Search 'ca':     {trie.search('ca')}")     # False (not a word!)
print(f"Search 'cards':  {trie.search('cards')}")  # False

print("\n--- starts_with() ---")
print(f"Prefix 'ca':  {trie.starts_with('ca')}")   # True
print(f"Prefix 'ba':  {trie.starts_with('ba')}")   # True
print(f"Prefix 'xyz': {trie.starts_with('xyz')}")  # False

print("\n--- autocomplete() ---")
print(f"'ca' → {trie.autocomplete('ca')}")         # cat, car, card, care
print(f"'ba' → {trie.autocomplete('ba')}")         # bat, ball, banana
print(f"'car'→ {trie.autocomplete('car')}")        # car, card, care

print("\n--- delete() ---")
trie.delete("card")
print(f"After deleting 'card':")
print(f"'car' suggestions: {trie.autocomplete('car')}")  # car, care (no card!)
````

````python
# ============================================
# CELL 1 — Graph (Adjacency List)
# ============================================
from collections import deque

class Graph:
    def __init__(self, directed=False):
        self.adjacency_list = {}
        self.directed = directed

    # Add vertex (node)
    def add_vertex(self, vertex):
        if vertex not in self.adjacency_list:
            self.adjacency_list[vertex] = []

    # Add edge (connection)
    def add_edge(self, v1, v2):
        self.add_vertex(v1)
        self.add_vertex(v2)
        self.adjacency_list[v1].append(v2)
        if not self.directed:               # undirected = both ways!
            self.adjacency_list[v2].append(v1)

    # Remove edge
    def remove_edge(self, v1, v2):
        self.adjacency_list[v1].remove(v2)
        if not self.directed:
            self.adjacency_list[v2].remove(v1)

    # Remove vertex
    def remove_vertex(self, vertex):
        del self.adjacency_list[vertex]
        for v in self.adjacency_list:       # remove all edges to this vertex
            if vertex in self.adjacency_list[v]:
                self.adjacency_list[v].remove(vertex)

    # Check if edge exists
    def has_edge(self, v1, v2):
        return v2 in self.adjacency_list.get(v1, [])

    # Get neighbors
    def neighbors(self, vertex):
        return self.adjacency_list.get(vertex, [])

    # BFS — Breadth First Search (level by level!)
    def bfs(self, start):
        visited = set()
        queue   = deque([start])
        result  = []
        visited.add(start)

        while queue:
            vertex = queue.popleft()       # dequeue front!
            result.append(vertex)

            for neighbor in self.adjacency_list[vertex]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor) # enqueue neighbors!
        return result

    # DFS — Depth First Search (go deep first!)
    def dfs(self, start):
        visited = set()
        result  = []

        def _dfs(vertex):
            visited.add(vertex)
            result.append(vertex)
            for neighbor in self.adjacency_list[vertex]:
                if neighbor not in visited:
                    _dfs(neighbor)         # go DEEP recursively!

        _dfs(start)
        return result

    # Check if path exists between two nodes
    def has_path(self, start, end):
        visited = set()

        def _dfs(vertex):
            if vertex == end:
                return True
            visited.add(vertex)
            for neighbor in self.adjacency_list[vertex]:
                if neighbor not in visited:
                    if _dfs(neighbor):
                        return True
            return False

        return _dfs(start)

    # Display graph
    def display(self):
        print("Graph (Adjacency List):")
        for vertex, neighbors in self.adjacency_list.items():
            print(f"  {vertex} → {neighbors}")


# ============================================
# TEST Graph!
# ============================================
print("=== UNDIRECTED GRAPH ===")
g = Graph()

# Add edges (vertices added automatically!)
edges = [("Nikhil","Priya"), ("Nikhil","Sara"),
         ("Priya","Raj"),    ("Priya","Divya"),
         ("Sara","Raj")]

for v1, v2 in edges:
    g.add_edge(v1, v2)

g.display()

print(f"\nNeighbors of Nikhil: {g.neighbors('Nikhil')}")
print(f"Has edge Nikhil→Raj: {g.has_edge('Nikhil','Raj')}")
print(f"Has edge Priya→Sara: {g.has_edge('Priya','Sara')}")

print(f"\nBFS from Nikhil: {g.bfs('Nikhil')}")
print(f"DFS from Nikhil: {g.dfs('Nikhil')}")

print(f"\nPath Nikhil→Divya: {g.has_path('Nikhil','Divya')}")
print(f"Path Nikhil→Karan: {g.has_path('Nikhil','Karan') if 'Karan' in g.adjacency_list else False}")
````
