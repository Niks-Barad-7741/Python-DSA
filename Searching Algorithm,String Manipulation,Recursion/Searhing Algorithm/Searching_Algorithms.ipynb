# 🔍 Searching Algorithms
> Find elements efficiently — from simple to advanced!

| Algorithm | Best | Average | Worst | Space | Requires Sorted? |
|-----------|------|---------|-------|-------|------------------|
| Linear Search | O(1) | O(n) | O(n) | O(1) | ❌ No |
| Binary Search | O(1) | O(log n) | O(log n) | O(1) | ✅ Yes |
| Ternary Search | O(1) | O(log₃ n) | O(log₃ n) | O(1) | ✅ Yes |
| Jump Search | O(1) | O(√n) | O(√n) | O(1) | ✅ Yes |
| Exponential Search | O(1) | O(log n) | O(log n) | O(1) | ✅ Yes |

## 1️⃣ Linear Search

**Idea:** Check EVERY element one by one until you find the target!

```
Array:  [64, 25, 12, 45, 11]
Target: 45

Step 1: 64 == 45? ❌
Step 2: 25 == 45? ❌
Step 3: 12 == 45? ❌
Step 4: 45 == 45? ✅ FOUND at index 3!
```

**When to use:**
- ✅ Unsorted array
- ✅ Small array
- ❌ Large sorted array → use Binary Search!

**Big O:** Best=O(1) | Average=O(n) | Worst=O(n) | Space=O(1)

````python
# ============================================
# Linear Search
# ============================================

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i           # return index!
    return -1                  # not found!


def linear_search_verbose(arr, target):
    print(f"Searching for {target} in {arr}")
    for i in range(len(arr)):
        print(f"  Step {i+1}: index[{i}] = {arr[i]}", end="")
        if arr[i] == target:
            print(f"  FOUND!")
            return i
        print(f" ")
    print(f"  Target {target} not found!")
    return -1


arr = [64, 25, 12, 45, 11, 78, 33]
print("=== LINEAR SEARCH ===")
print(f"Search 45 → index: {linear_search(arr, 45)}")
print(f"Search 99 → index: {linear_search(arr, 99)}")
print(f"Search 64 → index: {linear_search(arr, 64)}")
print()
linear_search_verbose(arr, 45)
````

## 2️⃣ Binary Search

**Idea:** Works ONLY on SORTED arrays! Cut the array in HALF each step!

```
Sorted: [11, 12, 25, 45, 64, 78, 99]
Target: 78

Step 1: left=0, right=6, mid=3 → arr[3]=45
        78 > 45 → go RIGHT!

Step 2: left=4, right=6, mid=5 → arr[5]=78
        78 == 78 ✅ FOUND at index 5!

Only 2 steps for 7 elements!
For 1,000,000 elements → only ~20 steps! ⚡
```

**Rule:**
```
arr[mid] == target → FOUND!
arr[mid]  < target → go RIGHT (left = mid + 1)
arr[mid]  > target → go LEFT  (right = mid - 1)
```

**Big O:** Best=O(1) | Average=O(log n) | Worst=O(log n) | Space=O(1)

````python
# ============================================
# Binary Search
# ============================================

def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = left + (right - left) // 2    # avoids overflow!

        if arr[mid] == target:
            return mid             # found!
        elif arr[mid] < target:
            left = mid + 1         # target bigger  → go RIGHT!
        else:
            right = mid - 1        # target smaller → go LEFT!

    return -1                      # not found!


def binary_search_verbose(arr, target):
    left, right = 0, len(arr) - 1
    step = 1
    print(f"Searching for {target} in {arr}")

    while left <= right:
        mid = left + (right - left) // 2
        print(f"  Step {step}: left={left}, right={right}, mid={mid} → arr[{mid}]={arr[mid]}", end="")

        if arr[mid] == target:
            print(f"  FOUND at index {mid}!")
            return mid
        elif arr[mid] < target:
            print(f" → go RIGHT!")
            left = mid + 1
        else:
            print(f" → go LEFT!")
            right = mid - 1
        step += 1

    print(f"  Target {target} not found!")
    return -1


arr = [11, 12, 25, 45, 64, 78, 99]    # MUST be sorted!
print("=== BINARY SEARCH ===")
print(f"Search 45 → index: {binary_search(arr, 45)}")
print(f"Search 99 → index: {binary_search(arr, 99)}")
print(f"Search 50 → index: {binary_search(arr, 50)}")
print()
binary_search_verbose(arr, 78)
````

## 3️⃣ Ternary Search

**Idea:** Like Binary Search but divides into THREE parts using TWO midpoints!

```
Sorted: [11, 12, 25, 45, 64, 78, 99]
Target: 78

Step 1:
  mid1 = left + (right-left)//3 = 0 + 2 = 2 → arr[2]=25
  mid2 = right - (right-left)//3 = 6 - 2 = 4 → arr[4]=64

  78 > arr[mid2]=64 → go RIGHT third!

Step 2: left=5, right=6
  mid1=5 → arr[5]=78 == 78 ✅ FOUND!
```

**3 regions:**
```
target < arr[mid1]  → LEFT third   (right = mid1-1)
target > arr[mid2]  → RIGHT third  (left  = mid2+1)
else                → MIDDLE third (left=mid1+1, right=mid2-1)
```

**Binary vs Ternary:**
```
Binary  → 1 comparison → eliminates 1/2
Ternary → 2 comparisons → eliminates 2/3
In practice: Binary is usually faster (fewer comparisons per step!)
Best use: Finding peak/min in unimodal functions!
```

**Big O:** Best=O(1) | Average=O(log₃ n) | Worst=O(log₃ n) | Space=O(1)

````python
# ============================================
# Ternary Search
# ============================================

def ternary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        third = (right - left) // 3
        mid1  = left  + third      # 1/3 point
        mid2  = right - third      # 2/3 point

        if arr[mid1] == target: return mid1
        if arr[mid2] == target: return mid2

        if target < arr[mid1]:
            right = mid1 - 1       # left third!
        elif target > arr[mid2]:
            left  = mid2 + 1       # right third!
        else:
            left  = mid1 + 1       # middle third!
            right = mid2 - 1

    return -1


def ternary_search_verbose(arr, target):
    left, right = 0, len(arr) - 1
    step = 1
    print(f"Searching for {target} in {arr}")

    while left <= right:
        third = (right - left) // 3
        mid1  = left  + third
        mid2  = right - third
        print(f"  Step {step}: mid1={mid1}({arr[mid1]}), mid2={mid2}({arr[mid2]})", end="")

        if arr[mid1] == target:
            print(f"  FOUND at mid1={mid1}!"); return mid1
        if arr[mid2] == target:
            print(f"  FOUND at mid2={mid2}!"); return mid2

        if target < arr[mid1]:
            print(f" → LEFT third!"); right = mid1 - 1
        elif target > arr[mid2]:
            print(f" → RIGHT third!"); left  = mid2 + 1
        else:
            print(f" → MIDDLE third!"); left = mid1 + 1; right = mid2 - 1
        step += 1

    print(f"  Target {target} not found!")
    return -1


arr = [11, 12, 25, 45, 64, 78, 99]
print("=== TERNARY SEARCH ===")
print(f"Search 78 → index: {ternary_search(arr, 78)}")
print(f"Search 11 → index: {ternary_search(arr, 11)}")
print(f"Search 50 → index: {ternary_search(arr, 50)}")
print()
ternary_search_verbose(arr, 78)
````

## 4️⃣ Jump Search

**Idea:** Jump in FIXED steps of √n, then linear search backwards!

```
Sorted: [11, 12, 25, 45, 64, 78, 99, 112, 130]
Target: 78
Step size = √9 = 3

JUMP phase:
  index 0→2  → arr[2]=25  < 78 → jump!
  index 3→5  → arr[5]=78  = 78 → jump!
  index 6→8  → arr[8]=130 > 78 → STOP! Overshot!

LINEAR phase (search back from index 6):
  arr[6]=99  > 78 → go back
  arr[5]=78 == 78 ✅ FOUND!
```

**Why √n is optimal step size:**
```
Too small step → too many jumps   (like linear!)
Too big  step → too much linear   (wasteful!)
√n steps → perfect balance!

Jump phase:   n/√n = √n jumps
Linear phase: √n   steps back
Total: √n + √n = O(√n)
```

**Big O:** Best=O(1) | Average=O(√n) | Worst=O(√n) | Space=O(1)

````python
# ============================================
# Jump Search
# ============================================
import math

def jump_search(arr, target):
    n    = len(arr)
    step = int(math.sqrt(n))    # optimal step = √n
    prev = 0

    # JUMP phase — jump forward until overshoot
    while arr[min(step, n) - 1] < target:
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return -1

    # LINEAR phase — search backwards
    while arr[prev] < target:
        prev += 1
        if prev == min(step, n):
            return -1

    if arr[prev] == target:
        return prev
    return -1


def jump_search_verbose(arr, target):
    n    = len(arr)
    step = int(math.sqrt(n))
    prev = 0
    print(f"Searching for {target} | n={n} | step=√{n}={step}")
    print(f"Array: {arr}")

    print(f"\n  JUMP phase:")
    while arr[min(step, n) - 1] < target:
        print(f"    arr[{min(step,n)-1}]={arr[min(step,n)-1]} < {target} → jump!")
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            print(f"  Not found!"); return -1

    print(f"    arr[{min(step,n)-1}]={arr[min(step,n)-1]} ≥ {target} → STOP! Linear back from index {prev}")

    print(f"  LINEAR phase:")
    while arr[prev] < target:
        print(f"    arr[{prev}]={arr[prev]} < {target} → move forward")
        prev += 1
        if prev == min(step, n):
            print(f"  Not found!"); return -1

    if arr[prev] == target:
        print(f"    arr[{prev}]={arr[prev]} == {target} ✅ FOUND at index {prev}!")
        return prev
    return -1


arr = [11, 12, 25, 45, 64, 78, 99, 112, 130]
print("=== JUMP SEARCH ===")
print(f"Search 78  → index: {jump_search(arr, 78)}")
print(f"Search 130 → index: {jump_search(arr, 130)}")
print(f"Search 50  → index: {jump_search(arr, 50)}")
print()
jump_search_verbose(arr, 78)
````

## 5️⃣ Exponential Search

**Idea:** DOUBLE the index until you overshoot, then Binary Search in that range!

```
Sorted: [11, 12, 25, 45, 64, 78, 99, 112, 130]
Target: 78

DOUBLING phase:
  index 1 → arr[1]=12  ≤ 78 → double to 2!
  index 2 → arr[2]=25  ≤ 78 → double to 4!
  index 4 → arr[4]=64  ≤ 78 → double to 8!
  index 8 → arr[8]=130 > 78 → STOP! Overshot!

Target is between index 4 and 8!

BINARY SEARCH phase in arr[4..8]:
  mid=6 → arr[6]=99 > 78 → go LEFT
  mid=5 → arr[5]=78 == 78 ✅ FOUND!
```

**Why useful?**
```
→ Great for UNBOUNDED arrays (size unknown!)
→ Great when target is near the START!
→ Same O(log n) as Binary Search but smarter range!
```

**Big O:** Best=O(1) | Average=O(log n) | Worst=O(log n) | Space=O(1)

````python
# ============================================
# Exponential Search
# ============================================

def binary_search_range(arr, target, left, right):    # helper with range!
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:   return mid
        elif arr[mid] < target:  left  = mid + 1
        else:                    right = mid - 1
    return -1


def exponential_search(arr, target):
    n = len(arr)
    if arr[0] == target:
        return 0

    i = 1
    while i < n and arr[i] <= target:
        i *= 2                              # double!

    left  = i // 2                          # range start
    right = min(i, n - 1)                   # range end
    return binary_search_range(arr, target, left, right)   # search only the range!


def exponential_search_verbose(arr, target):
    n = len(arr)
    print(f"Searching for {target} in {arr}")

    if arr[0] == target:
        print(f"  Found instantly at index 0! ")
        return 0

    i = 1
    print(f"  DOUBLING phase:")
    while i < n and arr[i] <= target:
        print(f"    index={i} → arr[{i}]={arr[i]} ≤ {target} → double to {i*2}!")
        i *= 2

    left  = i // 2
    right = min(i, n - 1)
    print(f"    index={i} → overshot! Binary search in range [{left}, {right}]")

    print(f"  BINARY SEARCH phase in arr[{left}..{right}]:")
    result = binary_search_range(arr, target, left, right)
    if result != -1:
        print(f"     FOUND at index {result}!")
    else:
        print(f"    Not found!")
    return result


arr = [11, 12, 25, 45, 64, 78, 99, 112, 130]
print("=== EXPONENTIAL SEARCH ===")
print(f"Search 78  → index: {exponential_search(arr, 78)}")
print(f"Search 11  → index: {exponential_search(arr, 11)}")
print(f"Search 50  → index: {exponential_search(arr, 50)}")
print()
exponential_search_verbose(arr, 78)
````

## ⚡ All 5 Algorithms — Side by Side Comparison!

````python
# ============================================
# All 5 Algorithms on same array!
# ============================================

arr = [11, 12, 25, 45, 64, 78, 99, 112, 130]
print(f"Array: {arr}")
print("=" * 55)

tests = [11, 78, 130, 50, 99]
for target in tests:
    l = linear_search(arr, target)
    b = binary_search(arr, target)
    t = ternary_search(arr, target)
    j = jump_search(arr, target)
    e = exponential_search(arr, target)

    status = f"index {l}" if l != -1 else "Not found"
    print(f"\nTarget = {target} → {status}")
    print(f"  Linear:      {l if l != -1 else 'Not found'}")
    print(f"  Binary:      {b if b != -1 else 'Not found'}")
    print(f"  Ternary:     {t if t != -1 else 'Not found'}")
    print(f"  Jump:        {j if j != -1 else 'Not found'}")
    print(f"  Exponential: {e if e != -1 else 'Not found'}")

print("\n" + "=" * 55)
print("\nBig O Summary:")
print(f"  {'Algorithm':22} {'Best':8} {'Worst':12} {'Sorted?'}")
print(f"  {'-'*52}")
print(f"  {'Linear Search':22} {'O(1)':8} {'O(n)':12} No")
print(f"  {'Binary Search':22} {'O(1)':8} {'O(log n)':12} Yes")
print(f"  {'Ternary Search':22} {'O(1)':8} {'O(log3 n)':12} Yes")
print(f"  {'Jump Search':22} {'O(1)':8} {'O(√n)':12} Yes")
print(f"  {'Exponential Search':22} {'O(1)':8} {'O(log n)':12} Yes")
````

## 🏆 Key Takeaways

### When to use which?
```
Unsorted array?              → Linear Search   (only option!)
Sorted array, general?       → Binary Search   (O(log n) always!)
Sorted, unimodal function?   → Ternary Search  (peak/min finding!)
Sorted, simple to implement? → Jump Search     (O(√n), easy!)
Unbounded/infinite array?    → Exponential     (unknown size!)
Target near the start?       → Exponential     (doubles fast!)
```

### Performance on n = 1,000,000 elements:
```
Linear Search      → up to 1,000,000 steps 😱
Jump Search        →          1,000 steps  (√1M)
Binary Search      →             20 steps  (log₂ 1M) ⚡
Ternary Search     →             13 steps  (log₃ 1M) ⚡
Exponential Search →             20 steps  (log₂ 1M) ⚡
```
