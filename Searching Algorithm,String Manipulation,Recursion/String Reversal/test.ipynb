```python
# String Reversal

def string_reversal(s):
    s=list(s)
    left = 0 
    right = len(s) - 1
    while left < right:
        s[left], s[right] = s[right], s[left]
        left += 1
        right -= 1
    return ''.join(s)
string_reversal("Hello, World!")
```

```python
# Reversing Word Without inbuilt functions

def reverse_word(s):
    words = s.split()
    left = 0 
    right = len(words) - 1
    while left < right:
        words[left], words[right] = words[right], words[left]
        left += 1
        right -= 1
    return ' '.join(words)
reverse_word("Hello World from Python")
```

```python
# Reverse word with inbuilt function

def reverse_word(s):
    word = s.split()
    word.reverse()
    return ' '.join(word)
reverse_word("Hello World from Python")
```

```python
# Rotation in string Left and right

def left_rotate_string(s,n):
    n = n % len(s)
    return s[n:] + s[:n]
def right_rotation(s,n):
    n = n % len(s)
    return s[-n:] + s[:-n]
def is_rotation(s1,s2):
    return len(s1) == len(s2) and s2 in (s1 + s1)


print(left_rotate_string("ABCDEF", 2))         # Output: CDEFAB
print(right_rotation("ABCDEF", 2))        # Output: EFABCD
print(is_rotation("ABCDEF", "CDEFAB"))  # Output: True
```

```python
# Removing Duplicate Characters from a String

def remove_duplicate(s):
    seen = set()
    result = []
    for char in s: 
        if char not in seen:
            seen.add(char)
            result.append(char)
    return ''.join(result)
print(remove_duplicate("programming"))
print(remove_duplicate("aabbccdd"))
```

```python
# most frequent character in a string

def most_frequent(s):
    frequency = {}
    for char in s:
        frequency[char] = frequency.get(char, 0)+1
    return max(frequency, key=frequency.get)

print(most_frequent("programming"))
print(most_frequent("aabbbbcc"))
```

```python
# anargam means to check the spelling of the word is same but the order of the character is different
# means listen and silent are anagram because they have same character but different order

def anagram_sort(s1,s2):
    return sorted(s1) == sorted(s2)

def anagram_frequency(s1,s2):
    freq = {}
    for char in s1:
        freq[char] = freq.get(char, 0) + 1
    for char in s2:
        freq[char] = freq.get(char, 0) - 1
        if freq[char] < 0:
            return False
    return True

print(anagram_sort("listen", "silent"))  # Output: True
print(anagram_sort("hello", "world"))    # Output: False
print(anagram_frequency("listen", "silent"))  # Output: True
print(anagram_frequency("hello", "world"))    # Output: False
    
    
```

```python
def palindrome(s):
    s = list(s)
    left = 0
    right= len(s)-1

    while left < right:
        if s[left]!= s[right]:
            return False
        left += 1
        right -= 1
    return True
print(palindrome("madam"))  # Output: True
print(palindrome("hello"))  # Output: False
```
