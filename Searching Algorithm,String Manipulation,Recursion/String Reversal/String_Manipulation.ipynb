## 1️⃣ Reversing a String

**Idea:** Use two pointers — one at start, one at end — swap and move inward!

```
s = ['H', 'e', 'l', 'l', 'o']
      ↑                   ↑
    left=0             right=4

Step 1: swap H ↔ o  → ['o', 'e', 'l', 'l', 'H']
Step 2: swap e ↔ l  → ['o', 'l', 'l', 'e', 'H']
Step 3: left=2, right=2 → left NOT < right → STOP!

Result: 'olleH' ✅
```

**Big O:** Time=O(n) | Space=O(n) — for the list conversion

````python
# ============================================
# 1. Reversing a String
# ============================================
# Technique : Two Pointer
# Big O     : Time O(n) | Space O(n)
# ============================================

# Method 1 — Two Pointer (manual, no built-ins)
def string_reversal(s):
    s     = list(s)
    left  = 0
    right = len(s) - 1
    while left < right:
        s[left], s[right] = s[right], s[left]
        left  += 1
        right -= 1
    return ''.join(s)


# Method 2 — Python slicing (one-liner!)
def string_reversal_slice(s):
    return s[::-1]


# Verbose — shows every swap!
def string_reversal_verbose(s):
    chars = list(s)
    left  = 0
    right = len(chars) - 1
    print(f"Original : {s}")
    print(f"As list  : {chars}")
    step = 1
    while left < right:
        print(f"  Step {step}: swap index[{left}]='{chars[left]}' ↔ index[{right}]='{chars[right]}'", end="")
        chars[left], chars[right] = chars[right], chars[left]
        left  += 1
        right -= 1
        print(f" → {chars}")
        step += 1
    result = ''.join(chars)
    print(f"Result   : {result} ✅")
    return result


print("=== STRING REVERSAL ===")
print(f"Two Pointer : {string_reversal('Hello, World!')}")
print(f"Slicing     : {string_reversal_slice('Hello, World!')}")
print(f"Single char : {string_reversal('a')}")
print(f"Empty       : {string_reversal('')}")
print()
string_reversal_verbose("Hello")
````

## 2️⃣ Reversing Words

**Idea:** Split sentence into words, then reverse the ORDER of words!

```
s = "Hello World from Python"

Step 1 — Split into words:
  ['Hello', 'World', 'from', 'Python']
    ↑                          ↑
  left=0                    right=3

Step 2 — Two pointer swap:
  swap Hello ↔ Python → ['Python', 'World', 'from', 'Hello']
  swap World ↔ from   → ['Python', 'from', 'World', 'Hello']

Step 3 — Join:
  'Python from World Hello' ✅

NOTE: Letters INSIDE each word stay the same!
Only the ORDER of words is reversed!
```

**Big O:** Time=O(n) | Space=O(n)

````python
# ============================================
# 2. Reversing Words
# ============================================
# Technique : Split + Two Pointer / Built-in
# Big O     : Time O(n) | Space O(n)
# ============================================

# Method 1 — Two Pointer (manual)
def reverse_words(s):
    words = s.split()
    left  = 0
    right = len(words) - 1
    while left < right:
        words[left], words[right] = words[right], words[left]
        left  += 1
        right -= 1
    return ' '.join(words)


# Method 2 — Built-in .reverse()
def reverse_words_builtin(s):
    words = s.split()
    words.reverse()
    return ' '.join(words)


# Method 3 — One-liner slicing
def reverse_words_slice(s):
    return ' '.join(s.split()[::-1])


# Verbose
def reverse_words_verbose(s):
    words = s.split()
    print(f"Original : '{s}'")
    print(f"Words    : {words}")
    left  = 0
    right = len(words) - 1
    step  = 1
    while left < right:
        print(f"  Step {step}: swap '{words[left]}' ↔ '{words[right]}'", end="")
        words[left], words[right] = words[right], words[left]
        left  += 1
        right -= 1
        print(f" → {words}")
        step += 1
    result = ' '.join(words)
    print(f"Result   : '{result}' ✅")
    return result


print("=== REVERSING WORDS ===")
s = "Hello World from Python"
print(f"Two Pointer : {reverse_words(s)}")
print(f"Built-in    : {reverse_words_builtin(s)}")
print(f"Slice       : {reverse_words_slice(s)}")
print()
reverse_words_verbose(s)
````

## 3️⃣ String Rotations

**Idea:** Shift characters left or right by n positions!

```
s = "ABCDEF"   n = 2

LEFT ROTATION by 2:
  Move first 2 chars to end!
  s[2:] + s[:2]  →  'CDEF' + 'AB'  →  'CDEFAB' ✅

RIGHT ROTATION by 2:
  Move last 2 chars to front!
  s[-2:] + s[:-2]  →  'EF' + 'ABCD'  →  'EFABCD' ✅

IS ROTATION check (s1='ABCDEF', s2='CDEFAB'):
  s1 + s1 = 'ABCDEFABCDEF'
  Is s2 inside 'ABCDEFABCDEF'? → YES! ✅
  This works because any rotation of s1
  will always appear in s1+s1!
```

**Why n % len(s)?**
```
Rotating by 8 on length-6 string = rotating by 2!
8 % 6 = 2  → prevents extra unnecessary rotations!
```

**Big O:** Time=O(n) | Space=O(n)

````python
# ============================================
# 3. Rotations
# ============================================
# Technique : Slicing + Concatenation
# Big O     : Time O(n) | Space O(n)
# ============================================

# Left rotation — move first n chars to end!
def left_rotate(s, n):
    n = n % len(s)          # handle n > len(s)!
    return s[n:] + s[:n]


# Right rotation — move last n chars to front!
def right_rotate(s, n):
    n = n % len(s)          # handle n > len(s)!
    return s[-n:] + s[:-n]


# Check if s2 is a rotation of s1!
def is_rotation(s1, s2):
    if len(s1) != len(s2):
        return False
    return s2 in (s1 + s1)  # genius trick!


# Verbose
def rotation_verbose(s, n):
    n_mod = n % len(s)
    print(f"Original       : '{s}'  (length={len(s)})")
    print(f"Rotation by    : {n} → effective = {n} % {len(s)} = {n_mod}")
    print()

    left = left_rotate(s, n)
    print(f"LEFT  ROTATE:")
    print(f"  s[{n_mod}:] = '{s[n_mod:]}'  +  s[:{n_mod}] = '{s[:n_mod]}'")
    print(f"  Result = '{left}' ✅")
    print()

    right = right_rotate(s, n)
    print(f"RIGHT ROTATE:")
    print(f"  s[-{n_mod}:] = '{s[-n_mod:]}'  +  s[:-{n_mod}] = '{s[:-n_mod]}'")
    print(f"  Result = '{right}' ✅")
    print()

    print(f"IS ROTATION check ('{left}' a rotation of '{s}'?):")
    print(f"  s1+s1 = '{s+s}'")
    print(f"  '{left}' in '{s+s}' → {is_rotation(s, left)} ✅")


print("=== ROTATIONS ===")
s = "ABCDEF"
print(f"Left  rotate by 2 : {left_rotate(s, 2)}")
print(f"Right rotate by 2 : {right_rotate(s, 2)}")
print(f"Left  rotate by 8 : {left_rotate(s, 8)}")
print(f"Is 'CDEFAB' rotation of 'ABCDEF'? {is_rotation('ABCDEF', 'CDEFAB')}")
print(f"Is 'ABCXYZ' rotation of 'ABCDEF'? {is_rotation('ABCDEF', 'ABCXYZ')}")
print()
rotation_verbose(s, 2)
````

## 4️⃣ Removing Duplicates

**Idea:** Use a set to track seen characters — only keep first occurrence!

```
s = 'programming'

char 'p' → not in seen → ADD  → seen={p}    result=[p]
char 'r' → not in seen → ADD  → seen={p,r}  result=[p,r]
char 'o' → not in seen → ADD  → seen={p,r,o} result=[p,r,o]
char 'g' → not in seen → ADD  → result=[p,r,o,g]
char 'r' → IN seen     → SKIP ❌
char 'a' → not in seen → ADD  → result=[p,r,o,g,a]
char 'm' → not in seen → ADD  → result=[p,r,o,g,a,m]
char 'm' → IN seen     → SKIP ❌
char 'i' → not in seen → ADD  → result=[p,r,o,g,a,m,i]
char 'n' → not in seen → ADD  → result=[p,r,o,g,a,m,i,n]
char 'g' → IN seen     → SKIP ❌

Result: 'programin' ✅
```

**Big O:** Time=O(n) | Space=O(k) where k = unique chars

````python
# ============================================
# 4. Removing Duplicates
# ============================================
# Technique : Set + List
# Big O     : Time O(n) | Space O(k)
# ============================================

def remove_duplicates(s):
    seen   = set()
    result = []
    for char in s:
        if char not in seen:
            seen.add(char)
            result.append(char)
    return ''.join(result)


# Verbose
def remove_duplicates_verbose(s):
    seen   = set()
    result = []
    print(f"Input : '{s}'")
    print(f"{'Char':6} {'In seen?':12} {'Action':8} {'Result so far'}")
    print("-" * 50)
    for char in s:
        if char not in seen:
            seen.add(char)
            result.append(char)
            print(f"  '{char}'   {'No':12} {'ADD ✅':8} {''.join(result)}")
        else:
            print(f"  '{char}'   {'Yes':12} {'SKIP ❌':8} {''.join(result)}")
    print(f"\nResult: '{''.join(result)}' ✅")
    return ''.join(result)


print("=== REMOVING DUPLICATES ===")
print(f"'programming' → '{remove_duplicates('programming')}'")
print(f"'aabbccdd'    → '{remove_duplicates('aabbccdd')}'")
print(f"'nikhil'      → '{remove_duplicates('nikhil')}'")
print(f"'aaaa'        → '{remove_duplicates('aaaa')}'")
print()
remove_duplicates_verbose("programming")
````

## 5️⃣ Most Repeated Character

**Idea:** Count frequency of every character using a hash map, return the max!

```
s = 'programming'

Step 1 — Build frequency map:
  { p:1, r:2, o:1, g:2, a:1, m:2, i:1, n:1 }

Step 2 — Find max:
  max by value → 'r', 'g', or 'm' all = 2
  max() returns first one found → 'r'

Result: 'r' ✅  (appears 2 times)
```

**Big O:** Time=O(n) | Space=O(k) where k = unique chars

````python
# ============================================
# 5. Most Repeated Character
# ============================================
# Technique : Hash Map + max()
# Big O     : Time O(n) | Space O(k)
# ============================================

def most_frequent(s):
    frequency = {}
    for char in s:
        frequency[char] = frequency.get(char, 0) + 1
    return max(frequency, key=frequency.get)


# Extended — return char AND count!
def most_frequent_with_count(s):
    frequency = {}
    for char in s:
        frequency[char] = frequency.get(char, 0) + 1
    best_char  = max(frequency, key=frequency.get)
    best_count = frequency[best_char]
    return best_char, best_count


# Verbose
def most_frequent_verbose(s):
    frequency = {}
    print(f"Input: '{s}'")
    print("\nBuilding frequency map:")
    for char in s:
        frequency[char] = frequency.get(char, 0) + 1
        print(f"  '{char}' → count = {frequency[char]}")
    print(f"\nFrequency map: {frequency}")
    best = max(frequency, key=frequency.get)
    print(f"Most frequent : '{best}' (appears {frequency[best]} times) ✅")
    return best


print("=== MOST REPEATED CHARACTER ===")
c, n = most_frequent_with_count("programming")
print(f"'programming' → '{c}' (appears {n} times)")
c, n = most_frequent_with_count("aabbbbcc")
print(f"'aabbbbcc'    → '{c}' (appears {n} times)")
c, n = most_frequent_with_count("nikhilbarad")
print(f"'nikhilbarad' → '{c}' (appears {n} times)")
print()
most_frequent_verbose("programming")
````

## 6️⃣ Anagrams

**Definition:** Two strings are anagrams if they have the **same characters** in **different order**!

```
'listen' and 'silent'
  l,i,s,t,e,n  =  e,i,l,n,s,t  ← same chars! ✅

'hello' and 'world'
  e,h,l,l,o  ≠  d,l,o,r,w     ← different! ❌
```

**Method 1 — Sort both:**
```
sorted('listen') = ['e','i','l','n','s','t']
sorted('silent') = ['e','i','l','n','s','t']
Equal? → YES → Anagram! ✅
Big O: O(n log n) — because of sorting!
```

**Method 2 — Frequency Map:**
```
Count s1 chars: {l:1, i:1, s:1, t:1, e:1, n:1}
Subtract s2 chars: decrement each
If any count goes below 0 → NOT anagram!
Big O: O(n) — faster! ⚡
```

````python
# ============================================
# 6. Anagrams
# ============================================
# Method 1: Sort — O(n log n)
# Method 2: Frequency Map — O(n)  ← faster!
# ============================================

# Method 1 — Sort both strings and compare
def anagram_sort(s1, s2):
    return sorted(s1.lower()) == sorted(s2.lower())


# Method 2 — Frequency map (faster, O(n))
def anagram_frequency(s1, s2):
    if len(s1) != len(s2):
        return False
    freq = {}
    for char in s1:
        freq[char] = freq.get(char, 0) + 1    # count s1!
    for char in s2:
        freq[char] = freq.get(char, 0) - 1    # subtract s2!
        if freq[char] < 0:
            return False                        # s2 has extra char!
    return True


# Verbose
def anagram_verbose(s1, s2):
    print(f"s1 = '{s1}'")
    print(f"s2 = '{s2}'")
    print()

    print(f"Method 1 — Sort:")
    sorted1 = sorted(s1.lower())
    sorted2 = sorted(s2.lower())
    print(f"  sorted(s1) = {sorted1}")
    print(f"  sorted(s2) = {sorted2}")
    print(f"  Equal?     = {sorted1 == sorted2} → {'Anagram! ✅' if sorted1==sorted2 else 'Not an anagram! ❌'}")
    print()

    print(f"Method 2 — Frequency Map:")
    freq = {}
    for char in s1:
        freq[char] = freq.get(char, 0) + 1
    print(f"  After counting s1: {freq}")
    failed = False
    for char in s2:
        freq[char] = freq.get(char, 0) - 1
        if freq[char] < 0:
            print(f"  '{char}' count went negative! ❌ Not an anagram!")
            failed = True
            break
    if not failed:
        print(f"  After subtracting s2: {freq}")
        print(f"  No negatives → Anagram! ✅")


print("=== ANAGRAMS ===")
pairs = [("listen", "silent"), ("hello", "world"),
         ("Astronomer", "Moon starer".replace(" ","")), ("abc", "ab")]
for s1, s2 in pairs:
    sort_r = anagram_sort(s1, s2)
    freq_r = anagram_frequency(s1, s2)
    print(f"'{s1}' vs '{s2}':  Sort={sort_r}  |  Frequency={freq_r}")
print()
anagram_verbose("listen", "silent")
````

## 7️⃣ Palindrome

**Definition:** A string that reads the same forwards AND backwards!

```
'madam'  → forwards: m-a-d-a-m
           backwards: m-a-d-a-m → SAME! ✅

'hello'  → forwards: h-e-l-l-o
           backwards: o-l-l-e-h → DIFFERENT! ❌
```

**Two Pointer approach:**
```
'madam'
  m  a  d  a  m
  ↑              ↑
left=0        right=4

Step 1: s[0]='m' == s[4]='m' ✅ → move inward
Step 2: s[1]='a' == s[3]='a' ✅ → move inward
Step 3: left=2, right=2 → left NOT < right → STOP!

All matched → Palindrome! ✅
```

**Big O:** Time=O(n) | Space=O(1) — most efficient!

````python
# ============================================
# 7. Palindrome
# ============================================
# Technique : Two Pointer
# Big O     : Time O(n) | Space O(1)
# ============================================

# Method 1 — Two Pointer (manual, O(1) space!)
def palindrome(s):
    left  = 0
    right = len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left  += 1
        right -= 1
    return True


# Method 2 — Slicing (one-liner, but O(n) space!)
def palindrome_slice(s):
    return s == s[::-1]


# Real world — ignore spaces and case!
def palindrome_clean(s):
    cleaned = ''.join(c.lower() for c in s if c.isalnum())
    return palindrome(cleaned)


# Verbose
def palindrome_verbose(s):
    chars = list(s)
    left  = 0
    right = len(chars) - 1
    print(f"Input: '{s}'")
    step = 1
    while left < right:
        match = chars[left] == chars[right]
        print(f"  Step {step}: s[{left}]='{chars[left]}' vs s[{right}]='{chars[right]}' → {'match ✅' if match else 'NO match ❌ → NOT palindrome!'}")
        if not match:
            return False
        left  += 1
        right -= 1
        step  += 1
    print(f"  All characters matched! → Palindrome ✅")
    return True


print("=== PALINDROME ===")
tests = ["madam", "hello", "racecar", "a", "",
         "A man a plan a canal Panama"]
for t in tests:
    two_ptr = palindrome(t)
    clean   = palindrome_clean(t)
    print(f"'{t}'")
    print(f"   Basic:     {two_ptr}")
    print(f"   Cleaned:   {clean}  ← ignores spaces/case")
print()
palindrome_verbose("madam")
print()
palindrome_verbose("hello")
````
