```python
# Linear Search Algorithm

def linear_search(arr,target):
    for i in range(len(arr)):
        if arr[i]==target:
            return ("Element found at index: "+str(i))
    return False

a=[1,8,14,6,45,9]
linear_search(a,99)
```

```python
# Binary Search Algorithm

A = [-3, -1, 0, 1, 4, 7]
def binary_search(arr,target):
    N=len(arr)
    L=0
    R=N-1

    while L<=R:
        M = L + ((R-L)//2)
        if arr[M]==target:
            return ("Element found at index: "+str(M))
        elif arr[M]<target:
            L = M+1
        else:
            R = M-1
    return False
binary_search(A,7)
```

```python
# Ternary Search Algorithm 

def ternary_search(arr,target):
    N=len(arr)
    L =0 
    R = N-1

    while L <=R:
        M1 = L + ((R-L)//3)
        M2 = R - ((R-L)//3)
        if arr[M1]==target:
            return ("Element found at index: "+str(M1))
        if arr[M2]==target:
            return ("Element found at index: "+str(M2))
        if target < arr[M1]:
            R = M1-1
        elif target > arr[M2]:
            L = M2+1
        else:
            L = M1+1
            R = M2-1
    return False
ternary_search(A,-1)
```

```python
# Jump Search Algorithm

import math 
def jump_search(arr,target):
    N = len(arr)
    step = int(math.sqrt(N))
    prev = 0 

    while arr[min(step,N)-1]<target:
        prev = step
        step += int(math.sqrt(N))
        if prev >= N:
            return False
    while arr[prev]<target:
        prev += 1
        if prev == min(step,N):
            return False
    if arr[prev]==target:
        return ("Element found at index: "+str(prev))
    return False

jump_search(A,99)
```
