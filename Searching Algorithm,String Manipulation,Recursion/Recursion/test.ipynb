```python
# factorial recursive function
# the factorial can be found in any number the formula is n! = n*(n-1)!
# in this recursive fuunction if we add two more zeros it will exceed the max recursion depth and it will give us an error but in python we can change the max recursion depth by using sys module and setrecursionlimit() function
def factorial(n):
    if n < 0: return "Negative number is not allowed"
    if n==0: return 1
    return n * factorial(n-1)
print(factorial(1000)) 
```

```python
# recursive fibonacci function
def fibonacci(n):
    if n < 0: return "Negative number is not allowed"
    if n==0: return 0
    if n==1: return 1
    return fibonacci(n-1) + fibonacci(n-2)

n = int(input("Enter the number to find Fibonacci sequence: "))
for i in range(n):
    print(fibonacci(i), end=" ")
```

```python
# Recursive Binary Search Tree 
def binarry_search(arr,target,L,R):
    if L > R: return -1
    M = L + ((R - L)//2)
    if arr[M] == target: return M
    if arr[M] > target: return binarry_search(arr,target,L,M-1)
    return binarry_search(arr,target,M+1,R)
arr = [1,2,3,4,5,6,7,8,9]
print(binarry_search(arr,9,0,len(arr)-1))
```

```python
# Recursive String Reversal
def reverse_string(s):
    if len(s) == 0: return s
    return s[-1] + reverse_string(s[:-1])

s = input("Enter a string to reverse: ")
print(reverse_string(s))
```

```python
# Recursive Palindrome Check
def is_palindrome_recursive(s, left, right):
    if left >= right: return True               
    if s[left] != s[right]: return False   
    return is_palindrome_recursive(s, left+1, right-1)

s = input("Enter a string to check if it's a palindrome: ")
print(is_palindrome_recursive(s, 0, len(s)-1))
```
