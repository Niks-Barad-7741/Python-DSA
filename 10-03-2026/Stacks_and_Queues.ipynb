# 📚 Stacks & Queues
> Two fundamental data structures based on how items are added and removed!

## 📚 PART 1 — Stack

### What is a Stack?

```
Stack = LIFO — Last In First Out!
Like a stack of plates!

Push →  [  30  ]  ← Pop
        [  20  ]
        [  10  ]

Last item added = First item removed!
```

### Real World Examples
- **Ctrl+Z (Undo)** → last action undone first!
- **Browser back button** → last page visited = first to go back!
- **Function call stack** → last function called = first to return!
- **Expression evaluation** → brackets matching!

### Stack Big O

| Operation | Big O | Description |
|-----------|-------|-------------|
| push() | O(1) ⚡ | Add to top |
| pop() | O(1) ⚡ | Remove from top |
| peek() | O(1) ⚡ | See top without removing |
| is_empty() | O(1) ⚡ | Check if empty |
| size() | O(1) ⚡ | Get number of elements |
| search() | O(n) | Find an element |
| clear() | O(1) | Remove all elements |

---

````python
# ============================================
# Stack — Full Implementation
# ============================================

class Stack:
    def __init__(self):
        self.stack = []       # list as underlying storage
        self._size = 0

    # O(1) — Push: add to top
    def push(self, value):
        self.stack.append(value)
        self._size += 1

    # O(1) — Pop: remove from top
    def pop(self):
        if self.is_empty():
            return "Stack Underflow! Stack is empty!"
        self._size -= 1
        return self.stack.pop()

    # O(1) — Peek: see top without removing
    def peek(self):
        if self.is_empty():
            return "Stack is empty!"
        return self.stack[-1]

    # O(1) — Check if empty
    def is_empty(self):
        return self._size == 0

    # O(1) — Get size
    def size(self):
        return self._size

    # O(n) — Search: find element, return position from top
    def search(self, value):
        for i in range(len(self.stack) - 1, -1, -1):   # search from top!
            if self.stack[i] == value:
                position = len(self.stack) - 1 - i      # position from top
                return f"Found '{value}' at position {position} from top!"
        return f"'{value}' not found in stack!"

    # O(1) — Clear: remove all elements
    def clear(self):
        self.stack = []
        self._size = 0
        print("Stack cleared!")

    # O(n) — Contains: check if value exists
    def contains(self, value):
        return value in self.stack

    # O(n) — Convert stack to list (top to bottom)
    def to_list(self):
        return self.stack[::-1]    # reverse = top to bottom!

    # Display
    def display(self):
        if self.is_empty():
            print("[ Empty Stack ]")
        else:
            print("Stack (top → bottom):")
            for i in range(len(self.stack) - 1, -1, -1):
                if i == len(self.stack) - 1:
                    print(f"  [ {self.stack[i]} ]  ← TOP")
                else:
                    print(f"  [ {self.stack[i]} ]")

    def __str__(self):
        return f"Stack{self.stack[::-1]}"
````

````python
# ============================================
# TEST — All Stack Methods
# ============================================
print("=" * 40)
print("      STACK — ALL METHODS")
print("=" * 40)

s = Stack()

# push()
print("\n--- push() ---")
s.push(10)
s.push(20)
s.push(30)
s.push(40)
s.display()

# peek()
print("\n--- peek() ---")
print(f"Top element: {s.peek()}")         # 40 (doesn't remove!)
print(f"Still there: {s.peek()}")         # still 40!

# size()
print("\n--- size() ---")
print(f"Size: {s.size()}")                # 4

# is_empty()
print("\n--- is_empty() ---")
print(f"Is empty: {s.is_empty()}")        # False

# search()
print("\n--- search() ---")
print(s.search(20))                        # Found at position 2 from top
print(s.search(99))                        # not found!

# contains()
print("\n--- contains() ---")
print(f"Contains 30: {s.contains(30)}")   # True
print(f"Contains 99: {s.contains(99)}")   # False

# to_list()
print("\n--- to_list() ---")
print(f"As list (top→bottom): {s.to_list()}")

# pop()
print("\n--- pop() ---")
print(f"Popped: {s.pop()}")               # 40 (LIFO!)
print(f"Popped: {s.pop()}")               # 30
s.display()

# clear()
print("\n--- clear() ---")
s.clear()
s.display()
print(f"Pop from empty: {s.pop()}")        # Stack Underflow!
````

## 🔄 PART 2 — Queue

### What is a Queue?

```
Queue = FIFO — First In First Out!
Like a line at a ticket counter!

Enqueue →  [ Nikhil | Priya | Sara ]  → Dequeue
            BACK                FRONT

First person in line = First person served!
```

### Real World Examples
- **Printer queue** → first document sent = first printed!
- **WhatsApp messages** → first sent = first received!
- **CPU scheduling** → first task in = first processed!
- **Customer support** → first ticket raised = first resolved!

### Queue Big O

| Operation | Big O | Description |
|-----------|-------|-------------|
| enqueue() | O(1) ⚡ | Add to back |
| dequeue() | O(1) ⚡ | Remove from front |
| peek() | O(1) ⚡ | See front without removing |
| peek_rear() | O(1) ⚡ | See back without removing |
| is_empty() | O(1) ⚡ | Check if empty |
| size() | O(1) ⚡ | Get number of elements |
| search() | O(n) | Find an element |
| clear() | O(1) | Remove all elements |
| contains() | O(n) | Check if value exists |

---

````python
# ============================================
# Queue — Full Implementation
# ============================================
from collections import deque    # faster than list!

class Queue:
    def __init__(self):
        self.queue = deque()      # deque = double ended queue
        self._size = 0

    # O(1) — Enqueue: add to back
    def enqueue(self, value):
        self.queue.append(value)  # add to RIGHT side
        self._size += 1

    # O(1) — Dequeue: remove from front
    def dequeue(self):
        if self.is_empty():
            return "Queue Underflow! Queue is empty!"
        self._size -= 1
        return self.queue.popleft()    # remove from LEFT side!

    # O(1) — Peek front: see front without removing
    def peek(self):
        if self.is_empty():
            return "Queue is empty!"
        return self.queue[0]           # first item!

    # O(1) — Peek rear: see back without removing
    def peek_rear(self):
        if self.is_empty():
            return "Queue is empty!"
        return self.queue[-1]          # last item!

    # O(1) — Check if empty
    def is_empty(self):
        return self._size == 0

    # O(1) — Get size
    def size(self):
        return self._size

    # O(n) — Search: find element, return position from front
    def search(self, value):
        for i, item in enumerate(self.queue):
            if item == value:
                return f"Found '{value}' at position {i} from front!"
        return f"'{value}' not found in queue!"

    # O(1) — Clear all
    def clear(self):
        self.queue.clear()
        self._size = 0
        print("Queue cleared!")

    # O(n) — Contains
    def contains(self, value):
        return value in self.queue

    # O(n) — Convert to list (front to back)
    def to_list(self):
        return list(self.queue)

    # Display
    def display(self):
        if self.is_empty():
            print("[ Empty Queue ]")
        else:
            items = list(self.queue)
            print("Queue (front → back):")
            print("  FRONT → " + " | ".join(str(x) for x in items) + " ← BACK")

    def __str__(self):
        return f"Queue{list(self.queue)}"
````

````python
# ============================================
# TEST — All Queue Methods
# ============================================
print("=" * 40)
print("      QUEUE — ALL METHODS")
print("=" * 40)

q = Queue()

# enqueue()
print("\n--- enqueue() ---")
q.enqueue("Nikhil")
q.enqueue("Priya")
q.enqueue("Sara")
q.enqueue("Raj")
q.display()

# peek() — front
print("\n--- peek() front ---")
print(f"Front: {q.peek()}")              # Nikhil (first in!)

# peek_rear() — back
print("\n--- peek_rear() back ---")
print(f"Back: {q.peek_rear()}")          # Raj (last in!)

# size()
print("\n--- size() ---")
print(f"Size: {q.size()}")              # 4

# is_empty()
print("\n--- is_empty() ---")
print(f"Is empty: {q.is_empty()}")      # False

# search()
print("\n--- search() ---")
print(q.search("Sara"))                  # position 2 from front
print(q.search("Karan"))                 # not found!

# contains()
print("\n--- contains() ---")
print(f"Contains Priya: {q.contains('Priya')}")  # True
print(f"Contains Karan: {q.contains('Karan')}")  # False

# to_list()
print("\n--- to_list() ---")
print(f"As list: {q.to_list()}")

# dequeue()
print("\n--- dequeue() ---")
print(f"Served: {q.dequeue()}")          # Nikhil (FIFO!)
print(f"Served: {q.dequeue()}")          # Priya
q.display()

# clear()
print("\n--- clear() ---")
q.clear()
q.display()
print(f"Dequeue from empty: {q.dequeue()}")  # Queue Underflow!
````

## 🌍 PART 3 — Real World Examples

````python
# ============================================
# Real World 1 — UNDO feature using Stack
# ============================================
print("=== UNDO FEATURE (Ctrl+Z) ===")

undo_stack = Stack()

# User types text step by step
actions = ["H", "He", "Hel", "Hell", "Hello", "Hello ", "Hello W", "Hello World"]
for action in actions:
    undo_stack.push(action)

print(f"Current text : '{undo_stack.peek()}'")
print(f"Total actions: {undo_stack.size()}")

print("\nUser presses Ctrl+Z three times:")
for i in range(3):
    undo_stack.pop()
    print(f"  After undo {i+1}: '{undo_stack.peek()}'")
````

````python
# ============================================
# Real World 2 — Bracket Matching using Stack
# ============================================
# Classic interview question!
# Check if brackets are balanced

def is_balanced(expression):
    stack = Stack()
    opening = "({["
    closing = ")}]"
    pairs  = {')': '(', '}': '{', ']': '['}

    for char in expression:
        if char in opening:
            stack.push(char)           # push opening bracket
        elif char in closing:
            if stack.is_empty():
                return False           # no matching opening!
            if stack.pop() != pairs[char]:
                return False           # wrong type!

    return stack.is_empty()            # all matched?

print("=== BRACKET MATCHING ===")
tests = [
    "()",
    "()[]{}",
    "([]{})",
    "(]",
    "([)]",
    "{[]}",
    "(((",
]

for test in tests:
    result = "✅ Balanced" if is_balanced(test) else "❌ Not balanced"
    print(f"  {test:15} → {result}")
````

````python
# ============================================
# Real World 3 — Printer Queue
# ============================================
print("=== PRINTER QUEUE ===")

printer = Queue()

# Documents sent to printer
documents = ["Resume.pdf", "Report.docx", "Invoice.xlsx", "Photo.png"]
for doc in documents:
    printer.enqueue(doc)
    print(f"Sent to printer: {doc}")

print()
printer.display()

print("\nPrinting...")
while not printer.is_empty():
    print(f"  ✅ Printed: {printer.dequeue()}")
````

````python
# ============================================
# Real World 4 — Browser History using Stack
# ============================================
print("=== BROWSER HISTORY ===")

history = Stack()
forward = Stack()    # for forward button!

def visit(url):
    history.push(url)
    forward.clear()  # clear forward history on new visit!
    print(f"Visiting: {url}")

def go_back():
    if history.size() <= 1:
        print("Can't go back!")
        return
    forward.push(history.pop())       # save current to forward!
    print(f"Back to: {history.peek()}")

def go_forward():
    if forward.is_empty():
        print("Can't go forward!")
        return
    page = forward.pop()
    history.push(page)
    print(f"Forward to: {page}")

visit("google.com")
visit("claude.ai")
visit("github.com")
visit("stackoverflow.com")

print()
go_back()       # github.com
go_back()       # claude.ai
go_forward()    # github.com
visit("youtube.com")  # new visit clears forward!
go_forward()    # can't go forward!
````

## ⚔️ PART 4 — Stack vs Queue Comparison

| Feature | Stack | Queue |
|---------|-------|-------|
| Full name | LIFO | FIFO |
| Meaning | Last In First Out | First In First Out |
| Add item | push() | enqueue() |
| Remove item | pop() | dequeue() |
| See next item | peek() — top | peek() — front |
| Add/Remove Big O | O(1) ⚡ | O(1) ⚡ |
| Real world | Undo, Back button | Printer, CPU tasks |

```
Stack → Think PLATES     → last plate placed = first removed!
Queue → Think LINE/QUEUE → first person in line = first served!
```

````python
# ============================================
# Exercise 1 — ANSWER
# ============================================

def reverse_string(text):
    s = Stack()
    for char in text:          # push each character
        s.push(char)
    result = ""
    while not s.is_empty():    # pop in reverse order!
        result += s.pop()
    return result

print(reverse_string("Nikhil"))    # lihkiN ✅
print(reverse_string("Hello"))     # olleH  ✅
````

## 🏆 Key Takeaways

1. **Stack = LIFO** → Last In First Out → like plates!
2. **Queue = FIFO** → First In First Out → like a line!
3. **Both add/remove in O(1)** → super fast!
4. **Stack use cases** → Undo, Back button, Brackets matching, Call stack
5. **Queue use cases** → Printer, CPU scheduling, Customer support
6. **deque** from collections is faster than list for Queue!
7. **Stack overflow** → too many pushes with no pop (memory full!)
8. **Underflow** → pop/dequeue from empty stack/queue!
