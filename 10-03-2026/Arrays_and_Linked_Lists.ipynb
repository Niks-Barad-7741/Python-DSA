# 🗂️ Arrays & 🔗 Linked Lists
> Two fundamental data structures with different strengths!

## 🗂️ PART 1 — Arrays

### What is an Array?

```
Array = items stored in CONTINUOUS memory!
Like seats in a cinema — all numbered, side by side!

Index:   0     1     2     3     4
       [ 10 ] [ 20 ] [ 30 ] [ 40 ] [ 50 ]
         ↑
    address 100  101  102  103  104

Python knows: address = start + index
So arr[3] → go to 100+3 = address 103 → instant! O(1)
```

### Array Big O

| Operation | Big O | Why? |
|-----------|-------|------|
| Access `arr[i]` | O(1) ⚡ | Direct address calculation |
| Search `x in arr` | O(n) | Check every element |
| Insert at end | O(1) ⚡ | Just add at back |
| Insert at start | O(n) 🐢 | Shift EVERYONE right |
| Delete at end | O(1) ⚡ | Just remove last |
| Delete at start | O(n) 🐢 | Shift EVERYONE left |

---

````python
# ============================================
# Array Operations
# ============================================
arr = [10, 20, 30, 40, 50]
print(f"Original:          {arr}")

# O(1) — Access by index
print(f"arr[2]:            {arr[2]}")        # 30

# O(1) — Insert at end
arr.append(60)
print(f"After append(60):  {arr}")

# O(n) — Insert at start (shifts everyone right!)
arr.insert(0, 5)
print(f"After insert(0,5): {arr}")

# O(n) — Delete from start (shifts everyone left!)
arr.pop(0)
print(f"After pop(0):      {arr}")

# O(1) — Delete from end
arr.pop()
print(f"After pop():       {arr}")

# O(n) — Search
print(f"30 in arr:         {30 in arr}")
print(f"99 in arr:         {99 in arr}")
````

## 🔗 PART 2 — Linked Lists

### What is a Linked List?

```
Linked List = items scattered ANYWHERE in memory!
Each item has VALUE + pointer to NEXT item!

HEAD
 ↓
[10 | •]──→ [20 | •]──→ [30 | •]──→ [40 | None]
address       address      address      address
  100           500          230          890

Items NOT next to each other in memory!
Connected only by pointers!
```

### Linked List Big O

| Operation | Big O | Why? |
|-----------|-------|------|
| Access by index | O(n) 🐢 | Follow pointers one by one |
| Search | O(n) | Check every node |
| Insert at start | O(1) ⚡ | Just update HEAD pointer! |
| Insert at end | O(n) 🐢 | Traverse to last node first |
| Delete at start | O(1) ⚡ | Just move HEAD forward! |
| Delete at end | O(n) 🐢 | Traverse to second last node |

---

````python
# ============================================
# Node — building block of Linked List
# ============================================
class Node:
    def __init__(self, value):
        self.value = value    # the data
        self.next = None      # pointer to next node

# ============================================
# Linked List class
# ============================================
class LinkedList:
    def __init__(self):
        self.head = None      # start of list
        self.size = 0

    # O(1) — Insert at start
    def insert_at_start(self, value):
        new_node = Node(value)
        new_node.next = self.head   # point to old head
        self.head = new_node        # new node is now head!
        self.size += 1

    # O(n) — Insert at end
    def insert_at_end(self, value):
        new_node = Node(value)
        if self.head is None:       # empty list!
            self.head = new_node
        else:
            current = self.head
            while current.next:     # go to last node
                current = current.next
            current.next = new_node # attach at end
        self.size += 1

    # O(1) — Delete from start
    def delete_from_start(self):
        if self.head is None:
            return "List is empty!"
        removed = self.head.value
        self.head = self.head.next  # move head forward!
        self.size -= 1
        return removed

    # O(n) — Search
    def search(self, value):
        current = self.head
        position = 0
        while current:
            if current.value == value:
                return f"Found {value} at position {position}!"
            current = current.next
            position += 1
        return f"{value} not found!"

    # Print all nodes
    def display(self):
        elements = []
        current = self.head
        while current:
            elements.append(str(current.value))
            current = current.next
        print(" → ".join(elements) + " → None")
````

````python
# ============================================
# TEST Linked List!
# ============================================
ll = LinkedList()

ll.insert_at_end(10)
ll.insert_at_end(20)
ll.insert_at_end(30)
ll.insert_at_start(5)    # goes to FRONT!

print("=== LINKED LIST ===")
ll.display()                              # 5 → 10 → 20 → 30 → None

print(ll.search(20))                      # Found 20 at position 2!
print(ll.search(99))                      # 99 not found!

print(f"Deleted: {ll.delete_from_start()}")  # removes 5
ll.display()                              # 10 → 20 → 30 → None
print(f"Size: {ll.size}")                 # 3
````

## ⚔️ PART 3 — Array vs Linked List

| Operation | Array | Linked List |
|-----------|-------|-------------|
| Memory | Continuous ✅ | Scattered |
| Access by index | O(1) ⚡ | O(n) 🐢 |
| Search | O(n) | O(n) |
| Insert at start | O(n) 🐢 | O(1) ⚡ |
| Insert at end | O(1) ⚡ | O(n) 🐢 |
| Delete at start | O(n) 🐢 | O(1) ⚡ |

```
Use ARRAY when       → you need fast ACCESS by index!
Use LINKED LIST when → you need fast INSERT/DELETE at start!
```

---

````python
# ============================================
# PART 4 — Side by side comparison
# ============================================
print("=== ARRAY ===")
arr = [10, 20, 30, 40, 50]
print(f"Access arr[3]:     {arr[3]}      → O(1) ⚡")
print(f"Insert at start:   inserting 5...")
arr.insert(0, 5)
print(f"Result: {arr}  → O(n) 🐢 (everyone shifted!)")

print()

print("=== LINKED LIST ===")
ll2 = LinkedList()
for val in [10, 20, 30, 40, 50]:
    ll2.insert_at_end(val)
ll2.display()
print("Insert 5 at start...")
ll2.insert_at_start(5)      # O(1) — nobody moved!
ll2.display()
print("→ O(1) ⚡ (only HEAD pointer updated!)")
````

---
## 🏆 Key Takeaways

1. **Array** → Fast ACCESS `O(1)`, Slow INSERT at start `O(n)`
2. **Linked List** → Fast INSERT at start `O(1)`, Slow ACCESS `O(n)`
3. **Array** → Memory is continuous (all together)
4. **Linked List** → Memory is scattered (connected by pointers)
5. **Use Array** → when you read data frequently
6. **Use Linked List** → when you insert/delete frequently at start
