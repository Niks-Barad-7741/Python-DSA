#  Big O Notation
> Measuring how fast or slow your code is as input size grows!

---

##  What is Big O?

Big O answers one question:

> **If my input grows, how do my steps grow?**

| Symbol | Name | Meaning |
|--------|------|---------|
| O(1) | Constant | Always same steps, no matter input size |
| O(log n) | Logarithmic | Cuts problem in half each time |
| O(n) | Linear | Steps grow with input size |
| O(n log n) | Linearithmic | Best sorting algorithms |
| O(n²) | Quadratic | Nested loops — gets slow fast! |
| O(2ⁿ) | Exponential | Avoid! Extremely slow |

---

## 
 Speed Comparison

For `n = 1000` elements:

```
O(1)       →           1 operation  ⚡ instant!
O(log n)   →          10 operations ✅ very fast!
O(n)       →       1,000 operations 🟡 ok
O(n log n) →      10,000 operations 🟠 acceptable
O(n²)      →   1,000,000 operations 🔴 slow!
O(2ⁿ)      → 2^1000 operations      💀 impossible!
```

---

##  How to Read Big O from Code

| What you SEE in code | What you WRITE |
|----------------------|---------------|
| `arr[i]` — direct access | O(1) |
| `x = 5 + 3` — math | O(1) |
| Single `for` loop | O(n) |
| Loop inside a loop | O(n²) |
| Cuts input in half each step | O(log n) |
| Two separate loops | O(n) — not O(2n)! |
| Nested loop + single loop | O(n²) — not O(n²+n)! |

## 1️⃣ O(1) — Constant Time

````python
# O(1) — Always 1 step, no matter the size!

def get_first(arr):
    return arr[0]      # always 1 step!

arr_small = [10, 20, 30]
arr_large = list(range(1_000_000))

print(get_first(arr_small))   # 10 — 1 step
print(get_first(arr_large))   # 0  — still 1 step!

# Other O(1) examples:
x = 5 + 3          # math
length = len(arr_large)  # Python stores length internally!
print(f"Length: {length}")
````

---
## 2️⃣ O(n) — Linear Time

````python
# O(n) — Steps grow with input size!

def find_name(employees, target):
    for emp in employees:       # checks every employee!
        if emp == target:
            return True
    return False

employees = ["Sara", "Priya", "Nikhil", "Raj", "Divya"]

print(find_name(employees, "Nikhil"))  # True
print(find_name(employees, "Karan"))   # False

# 5 employees   → max 5 checks
# 1000 employees → max 1000 checks
# Steps = n (size of input)
````

---
## 3️⃣ O(n²) — Quadratic Time

````python
# O(n²) — Nested loops! Gets slow very fast!

def find_duplicate(arr):
    for i in range(len(arr)):           # loop 1 → n times
        for j in range(len(arr)):       # loop 2 → n times
            if i != j and arr[i] == arr[j]:
                return arr[i]
    return None

arr = [1, 2, 3, 4, 2]
print(f"Duplicate: {find_duplicate(arr)}")  # 2

# n=10   →    100 operations
# n=100  →  10000 operations
# n=1000 → 1,000,000 operations 
````

---
## 4️⃣ O(log n) — Logarithmic Time

````python
# O(log n) — Cuts problem in HALF each time!
# Binary Search is the classic example!

def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2      # check middle!

        if arr[mid] == target:
            return f"Found {target} at index {mid}!"
        elif arr[mid] < target:
            left = mid + 1             # throw away LEFT half!
        else:
            right = mid - 1            # throw away RIGHT half!

    return f"{target} not found!"

arr = list(range(1000))     # 1000 elements
print(binary_search(arr, 756))

# 1000 elements → only ~10 steps!
# Because: 1000 → 500 → 250 → 125 → 62 → 31 → 15 → 7 → 3 → 1
````

##  The 2 Simplification Rules

### Rule 1 — Drop Constants
> `O(2n)` → `O(n)` &nbsp;&nbsp;&nbsp; `O(500)` → `O(1)`

### Rule 2 — Drop Smaller Terms  
> `O(n² + n)` → `O(n²)` &nbsp;&nbsp;&nbsp; `O(n + 100)` → `O(n)`

````python
# RULE 1 — Drop Constants
# Two separate loops = O(2n) → simplified to O(n)!

def two_loops(arr):
    for x in arr:    # O(n)
        print(x)
    for x in arr:    # O(n)
        print(x)
# Total = O(2n) → O(n) 
# The constant 2 doesn't matter at large scale!

print("Rule 1 — Drop Constants: O(2n) → O(n)")


# RULE 2 — Drop Smaller Terms
# Nested loop + single loop = O(n² + n) → O(n²)!

def nested_plus_single(arr):
    for i in arr:        # O(n²) ← dominates!
        for j in arr:
            print(i, j)
    for x in arr:        # O(n)  ← dropped!
        print(x)
# Total = O(n² + n) → O(n²) 
# n is tiny compared to n² at large scale!

print("Rule 2 — Drop Smaller Terms: O(n² + n) → O(n²)")
````

##  Quick Reference Cheat Sheet

| Code Pattern | Big O | Name |
|-------------|-------|------|
| `arr[0]`, `arr[i]` | O(1) | Constant |
| `x = 5 + 3` | O(1) | Constant |
| `len(arr)` | O(1) | Constant |
| Single `for` loop | O(n) | Linear |
| Two separate loops | O(n) | Linear |
| Loop inside loop | O(n²) | Quadratic |
| Binary search | O(log n) | Logarithmic |
| `arr.sort()` | O(n log n) | Linearithmic |

---

##  Key Rules to Remember

1. **Drop constants** → `O(2n)` = `O(n)`
2. **Drop smaller terms** → `O(n² + n)` = `O(n²)`
3. **No loop** → `O(1)`
4. **One loop** → `O(n)`
5. **Loop inside loop** → `O(n²)`
6. **Half input each time** → `O(log n)`
