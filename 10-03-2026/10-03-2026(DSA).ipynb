```python
# ================================
# O(1) — Constant Time
# ================================
# Speed does NOT depend on input size!

def get_first(arr):
    return arr[0]      # always 1 step, no matter size!

arr = [10, 20, 30, 40, 50]
print(get_first(arr))  # 10 — 1 step

arr = list(range(1000000))
print(get_first(arr))  # still 1 step! ⚡


# ================================
# O(n) — Linear Time
# ================================
# Speed grows with input size!

def find_name(employees, target):
    for emp in employees:      # checks EVERY employee!
        if emp == target:
            return True
    return False

# 10 employees   → max 10 checks
# 1000 employees → max 1000 checks
# Speed grows linearly with n!


# ================================
# O(n²) — Quadratic Time
# ================================
# Nested loops! Very slow for large inputs!

def find_duplicate(arr):
    for i in range(len(arr)):           # loop 1 → n times
        for j in range(len(arr)):       # loop 2 → n times
            if i != j and arr[i] == arr[j]:
                return arr[i]

# n=10   →   100 operations
# n=100  → 10000 operations
# n=1000 → 1,000,000 operations 🔴


# ================================
# O(log n) — Logarithmic Time
# ================================
# Cuts problem in HALF each time!
# Binary search is classic example!

def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2      # check middle!

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1             # throw away LEFT half!
        else:
            right = mid - 1            # throw away RIGHT half!

    return -1

arr = list(range(1000))
print(binary_search(arr, 756))
# 1000 elements → only ~10 steps! ⚡
# Because: 1000 → 500 → 250 → 125 → 62 → 31 → 15 → 7 → 3 → 1


# ================================
# O(n log n) — Best sorting!
# ================================
arr = [5, 2, 8, 1, 9, 3]
arr.sort()       # Python's built-in sort = O(n log n)
print(arr)       # [1, 2, 3, 5, 8, 9]
```

```python
# RULE 1 — Drop constants!
# O(2n) → O(n)
# O(500) → O(1)

def example(arr):
    for x in arr:        # O(n)
        print(x)
    for x in arr:        # O(n)
        print(x)
# Total = O(2n) → simplified to O(n)!


# RULE 2 — Drop smaller terms!
# O(n² + n) → O(n²)
# n² dominates when n is large!

def example2(arr):
    for i in arr:            # O(n)
        for j in arr:        # O(n)
            print(i, j)      # → O(n²)
    for x in arr:            # O(n)
        print(x)
# Total = O(n² + n) → simplified to O(n²)!


# RULE 3 — Different inputs = different variables!
def example3(arr1, arr2):
    for x in arr1:       # O(a)
        print(x)
    for y in arr2:       # O(b)
        print(y)
# Total = O(a + b)   NOT O(n)!
# Because arr1 and arr2 are different sizes!
```
