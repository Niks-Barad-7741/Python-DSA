# #️⃣ Hash Tables
> Store and retrieve data in O(1) using Key → Value pairs!

---

## 🤔 What is a Hash Table?

```
Hash Table = stores data as KEY → VALUE pairs!
Uses a HASH FUNCTION to decide WHERE to store each value!

Example:
  'Nikhil' → 95
  'Priya'  → 87
  'Sara'   → 91

You give a KEY → instantly get the VALUE! O(1) ⚡
```

### How it works internally:
```
Step 1 → You give key:           'Nikhil'
Step 2 → Hash function runs:     hash('Nikhil') → 3
Step 3 → Value stored at index:  array[3] = 95
Step 4 → To retrieve:            hash('Nikhil') → 3 → 95 ⚡
```

### What is a Collision?
```
Collision = two keys get the SAME index!
Solution  = Chaining (store multiple values at same index)

Index 7: [Nikhil:95] → [Priya:87] → None
```

### Hash Table Big O

| Operation | Average | Worst Case |
|-----------|---------|------------|
| Insert | O(1) ⚡ | O(n) |
| Search | O(1) ⚡ | O(n) |
| Delete | O(1) ⚡ | O(n) |
| Access | O(1) ⚡ | O(n) |

> Average is almost always O(1)! Worst case only if ALL keys collide (very rare!)

````python
# ============================================
# CELL 1 — Python dict IS a Hash Table!
# ============================================
print("=== Python dict = Hash Table ===")

student_marks = {}

# Insert — O(1)
student_marks["Nikhil"] = 95
student_marks["Priya"]  = 87
student_marks["Sara"]   = 91
student_marks["Raj"]    = 78
print(f"All data:     {student_marks}")

# Access — O(1)
print(f"Nikhil marks: {student_marks['Nikhil']}")

# Update — O(1)
student_marks["Nikhil"] = 98
print(f"After update: {student_marks['Nikhil']}")

# Delete — O(1)
del student_marks["Raj"]
print(f"After delete: {student_marks}")

# Search key — O(1)
print(f"'Priya' exists: {'Priya' in student_marks}")
print(f"'Raj' exists:   {'Raj' in student_marks}")

# Get with default — safe access!
print(f"Karan marks: {student_marks.get('Karan', 'Not found!')}")
````

````python
# ============================================
# CELL 2 — All dict Methods!
# ============================================
print("=== ALL HASH TABLE METHODS ===\n")

employee = {
    "name"      : "Nikhil",
    "department": "Finance",
    "salary"    : 156248,
    "city"      : "Bangalore",
    "remote"    : True
}

# keys()
print("--- keys() ---")
print(list(employee.keys()))

# values()
print("\n--- values() ---")
print(list(employee.values()))

# items()
print("\n--- items() ---")
for key, value in employee.items():
    print(f"  {key:12} → {value}")

# update()
print("\n--- update() ---")
employee.update({"bonus": 15000, "salary": 160000})
print(f"Salary updated: {employee['salary']}")
print(f"Bonus added:    {employee['bonus']}")

# pop()
print("\n--- pop() ---")
city = employee.pop("city")
print(f"Removed city: {city}")

# get()
print("\n--- get() ---")
print(f"Name:  {employee.get('name', 'Unknown')}")
print(f"Phone: {employee.get('phone', 'Not provided')}")

# len()
print("\n--- len() ---")
print(f"Total keys: {len(employee)}")

# in operator
print("\n--- 'in' operator ---")
print(f"'name' in dict:  {'name' in employee}")
print(f"'phone' in dict: {'phone' in employee}")

# clear()
print("\n--- clear() ---")
temp = {"a": 1, "b": 2}
temp.clear()
print(f"After clear: {temp}")
````

````python
# ============================================
# CELL 3 — Build Hash Table from Scratch!
# ============================================

class HashTable:
    def __init__(self, size=10):
        self.size = size
        self.table = [[] for _ in range(size)]  # chaining!

    # Hash function — converts key to index
    def _hash(self, key):
        return sum(ord(c) for c in str(key)) % self.size

    # O(1) avg — Insert / Update
    def insert(self, key, value):
        index = self._hash(key)
        for pair in self.table[index]:
            if pair[0] == key:
                pair[1] = value        # update existing!
                return
        self.table[index].append([key, value])  # new key!

    # O(1) avg — Get value
    def get(self, key):
        index = self._hash(key)
        for pair in self.table[index]:
            if pair[0] == key:
                return pair[1]
        return "Key not found!"

    # O(1) avg — Delete
    def delete(self, key):
        index = self._hash(key)
        for i, pair in enumerate(self.table[index]):
            if pair[0] == key:
                self.table[index].pop(i)
                return f"Deleted '{key}'!"
        return "Key not found!"

    # O(1) avg — Check if key exists
    def contains(self, key):
        index = self._hash(key)
        for pair in self.table[index]:
            if pair[0] == key:
                return True
        return False

    # Display entire hash table
    def display(self):
        print("Hash Table internals:")
        for i, bucket in enumerate(self.table):
            if bucket:
                print(f"  [{i}] → {bucket}")


# TEST!
print("=== CUSTOM HASH TABLE ===")
ht = HashTable()

ht.insert("Nikhil", 95)
ht.insert("Priya",  87)
ht.insert("Sara",   91)
ht.insert("Raj",    78)
ht.display()

print(f"\nGet Nikhil:     {ht.get('Nikhil')}")
print(f"Get Karan:      {ht.get('Karan')}")
print(f"Contains Priya: {ht.contains('Priya')}")

ht.insert("Nikhil", 100)    # update!
print(f"\nUpdated Nikhil: {ht.get('Nikhil')}")

print(ht.delete("Raj"))
ht.display()
````

````python
# ============================================
# CELL 4 — Real World Examples
# ============================================

# 1. Word frequency counter
print("=== WORD FREQUENCY ===")
text = "nikhil is learning python nikhil is good at python"
words = text.split()
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

for word, count in sorted(frequency.items(), key=lambda x: x[1], reverse=True):
    print(f"  {word:10} → {count} times")


# 2. Employee lookup system
print("\n=== EMPLOYEE LOOKUP ===")
employees = {
    "EMP1000": {"name": "Nikhil", "dept": "Finance", "salary": 156248},
    "EMP1001": {"name": "Priya",  "dept": "HR",      "salary": 93021},
    "EMP1002": {"name": "Sara",   "dept": "IT",       "salary": 89286},
}

emp_id = "EMP1000"
emp = employees.get(emp_id, "Not found!")
print(f"Looking up {emp_id}: {emp['name']} — {emp['dept']} — {emp['salary']}")


# 3. Cache using hash table
print("\n=== SIMPLE CACHE ===")
cache = {}

def get_data(key):
    if key in cache:
        print(f"Cache HIT!  → {cache[key]}")
        return cache[key]
    print(f"Cache MISS! → fetching...")
    result = f"data_for_{key}"
    cache[key] = result
    return result

get_data("user_1")    # miss
get_data("user_2")    # miss
get_data("user_1")    # HIT!
get_data("user_2")    # HIT!
````

## 🏆 Key Takeaways

1. **Hash Table** = Key → Value storage with O(1) access!
2. **Hash Function** = converts key to array index!
3. **Collision** = two keys get same index → solved by chaining!
4. **Python dict** is a built-in Hash Table!
5. **Best use cases** → lookups, frequency counting, caching!
6. **Average O(1)** for insert, search, delete!
7. **Keys must be unique** → inserting same key updates value!

### All dict methods summary:

| Method | Description |
|--------|-------------|
| `d[key] = val` | Insert / Update |
| `d[key]` | Access |
| `d.get(key, default)` | Safe access |
| `del d[key]` | Delete |
| `d.pop(key)` | Delete and return |
| `key in d` | Check existence |
| `d.keys()` | All keys |
| `d.values()` | All values |
| `d.items()` | All key-value pairs |
| `d.update({...})` | Merge another dict |
| `d.clear()` | Remove all |
| `len(d)` | Number of keys |
