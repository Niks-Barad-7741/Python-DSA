# 🔃 Sorting Algorithms
> Seven sorting algorithms — from simple to advanced!

| Algorithm | Best | Average | Worst | Space | Stable? |
|-----------|------|---------|-------|-------|---------|
| Bubble Sort | O(n) | O(n²) | O(n²) | O(1) | ✅ |
| Selection Sort | O(n²) | O(n²) | O(n²) | O(1) | ❌ |
| Insertion Sort | O(n) | O(n²) | O(n²) | O(1) | ✅ |
| Merge Sort | O(n log n) | O(n log n) | O(n log n) | O(n) | ✅ |
| Quick Sort | O(n log n) | O(n log n) | O(n²) | O(log n) | ❌ |
| Counting Sort | O(n+k) | O(n+k) | O(n+k) | O(k) | ✅ |
| Bucket Sort | O(n+k) | O(n+k) | O(n²) | O(n) | ✅ |

> **Stable** = equal elements keep their original order!

## 1️⃣ Bubble Sort

**Idea:** Compare adjacent elements, swap if wrong order. Largest element *bubbles* to the end each pass!

```
Pass 1: [64, 25, 12, 45, 11]
  64>25 → swap → [25, 64, 12, 45, 11]
  64>12 → swap → [25, 12, 64, 45, 11]
  64>45 → swap → [25, 12, 45, 64, 11]
  64>11 → swap → [25, 12, 45, 11, 64] ← 64 in place! ✅
```

**Big O:** O(n²) average | O(n) best (already sorted) | O(1) space

````python
# ============================================
# Bubble Sort
# ============================================

def bubble_sort(arr):
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):    # last i elements already sorted!
            if arr[j] > arr[j + 1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:                   # no swaps = already sorted!
            break
    return arr


def bubble_sort_verbose(arr):
    arr = arr.copy()
    n = len(arr)
    print("Step by step:")
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        print(f"  Pass {i+1}: {arr}")
        if not swapped:
            print("  No swaps — already sorted! Stopping early!")
            break
    return arr


arr = [64, 25, 12, 45, 11]
print("=== BUBBLE SORT ===")
print(f"Original: {arr}")
print(f"Sorted:   {bubble_sort(arr)}")
print()
bubble_sort_verbose(arr)
````

## 2️⃣ Selection Sort

**Idea:** Find the minimum element, put it at the start. Repeat for remaining array!

```
Pass 1: Find min in [64,25,12,45,11] → min=11 → swap with 64
        [11, 25, 12, 45, 64] ✅
Pass 2: Find min in [25,12,45,64]    → min=12 → swap with 25
        [11, 12, 25, 45, 64] ✅
```

**Big O:** O(n²) always | O(1) space | NOT stable

````python
# ============================================
# Selection Sort
# ============================================

def selection_sort(arr):
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):          # find minimum in remaining!
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]   # swap!
    return arr


def selection_sort_verbose(arr):
    arr = arr.copy()
    n = len(arr)
    print("Step by step:")
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        print(f"  Pass {i+1}: {arr}  ← {arr[i]} placed!")
    return arr


arr = [64, 25, 12, 45, 11]
print("=== SELECTION SORT ===")
print(f"Original: {arr}")
print(f"Sorted:   {selection_sort(arr)}")
print()
selection_sort_verbose(arr)
````

## 3️⃣ Insertion Sort

**Idea:** Like sorting playing cards! Pick each element and insert it into the correct position in the sorted part!

```
[64] → pick 25 → 25<64 → shift 64 → [25, 64]
     → pick 12 → shift 64,25       → [12, 25, 64]
     → pick 45 → shift 64          → [12, 25, 45, 64]
     → pick 11 → shift all         → [11, 12, 25, 45, 64] ✅
```

**Big O:** O(n²) average | O(n) best | O(1) space | Stable ✅

````python
# ============================================
# Insertion Sort
# ============================================

def insertion_sort(arr):
    arr = arr.copy()
    for i in range(1, len(arr)):
        key = arr[i]           # element to insert!
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]  # shift right!
            j -= 1
        arr[j+1] = key         # insert in correct position!
    return arr


def insertion_sort_verbose(arr):
    arr = arr.copy()
    print("Step by step:")
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
        print(f"  Insert {key:2}: {arr}")
    return arr


arr = [64, 25, 12, 45, 11]
print("=== INSERTION SORT ===")
print(f"Original: {arr}")
print(f"Sorted:   {insertion_sort(arr)}")
print()
insertion_sort_verbose(arr)
````

## 4️⃣ Merge Sort

**Idea:** Divide & Conquer! Split array in half → sort each half → merge!

```
DIVIDE:  [64, 25, 12, 45, 11]
         [64, 25]    [12, 45, 11]
         [64][25]    [12][45][11]

MERGE:   [25, 64]    [11, 12, 45]
         [11, 12, 25, 45, 64] ✅
```

**Big O:** O(n log n) always | O(n) space | Stable ✅

````python
# ============================================
# Merge Sort
# ============================================

def merge_sort(arr):
    if len(arr) <= 1:
        return arr                        # base case!
    mid   = len(arr) // 2
    left  = merge_sort(arr[:mid])         # sort left!
    right = merge_sort(arr[mid:])         # sort right!
    return merge(left, right)             # merge both!


def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i]); i += 1
        else:
            result.append(right[j]); j += 1
    result.extend(left[i:])              # remaining elements!
    result.extend(right[j:])
    return result


arr = [64, 25, 12, 45, 11]
print("=== MERGE SORT ===")
print(f"Original: {arr}")
print(f"Sorted:   {merge_sort(arr)}")

# Show divide steps
print("\nDivide steps:")
def merge_sort_verbose(arr, depth=0):
    indent = "  " * depth
    if len(arr) <= 1:
        print(f"{indent}Base: {arr}")
        return arr
    mid = len(arr) // 2
    print(f"{indent}Split: {arr} → {arr[:mid]} | {arr[mid:]}")
    left  = merge_sort_verbose(arr[:mid],  depth+1)
    right = merge_sort_verbose(arr[mid:], depth+1)
    merged = merge(left, right)
    print(f"{indent}Merge: {left} + {right} → {merged}")
    return merged

merge_sort_verbose([64, 25, 12, 45, 11])
````

## 5️⃣ Quick Sort

**Idea:** Pick a PIVOT → smaller elements left, larger right → repeat!

```
Array: [64, 25, 12, 45, 11]  pivot = 45

Left  (< 45): [25, 12, 11]
Pivot:        [45]
Right (> 45): [64]

Sort each side → [11,12,25] + [45] + [64]
= [11, 12, 25, 45, 64] ✅
```

**Big O:** O(n log n) average | O(n²) worst | O(log n) space | NOT stable

````python
# ============================================
# Quick Sort
# ============================================

def quick_sort(arr):
    if len(arr) <= 1:
        return arr                            # base case!
    pivot = arr[len(arr) // 2]               # pick middle as pivot!
    left  = [x for x in arr if x < pivot]   # smaller than pivot
    mid   = [x for x in arr if x == pivot]  # equal to pivot
    right = [x for x in arr if x > pivot]   # larger than pivot
    return quick_sort(left) + mid + quick_sort(right)


def quick_sort_verbose(arr, depth=0):
    indent = "  " * depth
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left  = [x for x in arr if x < pivot]
    mid   = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    print(f"{indent}Array={arr} | Pivot={pivot} | Left={left} | Right={right}")
    return quick_sort_verbose(left, depth+1) + mid + quick_sort_verbose(right, depth+1)


arr = [64, 25, 12, 45, 11]
print("=== QUICK SORT ===")
print(f"Original: {arr}")
print(f"Sorted:   {quick_sort(arr)}")
print("\nStep by step:")
quick_sort_verbose(arr)
````

## 6️⃣ Counting Sort

**Idea:** Count occurrences of each number, then rebuild sorted array!

```
Array: [4, 2, 2, 8, 3, 3, 1]

Count: index=0,1,2,3,4,5,6,7,8
       count=0,1,2,2,1,0,0,0,1

Rebuild: 1×1, 2×2, 2×3, 1×4, 1×8
Result: [1, 2, 2, 3, 3, 4, 8] ✅
```

**Big O:** O(n+k) | O(k) space | Only works for integers! | Stable ✅

````python
# ============================================
# Counting Sort
# ============================================

def counting_sort(arr):
    if not arr: return arr
    max_val = max(arr)
    count   = [0] * (max_val + 1)     # count array!
    for num in arr:
        count[num] += 1               # count each number!
    result = []
    for i, c in enumerate(count):
        result.extend([i] * c)        # rebuild sorted!
    return result


arr = [4, 2, 2, 8, 3, 3, 1]
print("=== COUNTING SORT ===")
print(f"Original: {arr}")
print(f"Sorted:   {counting_sort(arr)}")

# Show count array
max_val = max(arr)
count   = [0] * (max_val + 1)
for num in arr: count[num] += 1
print(f"\nCount array:")
print(f"  Index: {list(range(max_val+1))}")
print(f"  Count: {count}")
print(f"\nReading count array:")
for i, c in enumerate(count):
    if c > 0:
        print(f"  {i} appears {c} time(s)")
````

## 7️⃣ Bucket Sort

**Idea:** Divide elements into buckets → sort each bucket → combine!

```
Array: [0.42, 0.32, 0.73, 0.11, 0.85, 0.56]

Bucket 0 [0.0-0.2]: [0.11]
Bucket 1 [0.2-0.4]: [0.32, 0.42]
Bucket 2 [0.4-0.6]: [0.56]
Bucket 3 [0.6-0.8]: [0.73]
Bucket 4 [0.8-1.0]: [0.85]

Result: [0.11, 0.32, 0.42, 0.56, 0.73, 0.85] ✅
```

**Big O:** O(n+k) average | O(n²) worst | O(n) space | Stable ✅

````python
# ============================================
# Bucket Sort
# ============================================

def bucket_sort(arr, show_buckets=False):
    if not arr: return arr
    n            = len(arr)
    max_val      = max(arr)
    min_val      = min(arr)
    bucket_range = (max_val - min_val) / n + 1
    buckets      = [[] for _ in range(n)]

    for num in arr:
        idx = int((num - min_val) / bucket_range)
        if idx >= n: idx = n - 1
        buckets[idx].append(num)

    result = []
    for i, bucket in enumerate(buckets):
        bucket.sort()
        if bucket and show_buckets:
            print(f"  Bucket {i}: {bucket}")
        result.extend(bucket)
    return result


arr = [0.42, 0.32, 0.73, 0.11, 0.85, 0.56]
print("=== BUCKET SORT ===")
print(f"Original: {arr}")
print(f"\nBuckets:")
sorted_arr = bucket_sort(arr, show_buckets=True)
print(f"\nSorted:   {sorted_arr}")

# Integer array example
print("\n--- Integer array ---")
arr2 = [64, 25, 12, 45, 11, 78, 33]
print(f"Original: {arr2}")
print(f"Sorted:   {bucket_sort(arr2)}")
````

````python
# ============================================
# ALL Algorithms — Side by Side Comparison
# ============================================

arr = [64, 25, 12, 45, 78, 11, 99, 33, 57, 22]
print(f"Original: {arr}")
print("=" * 50)

algorithms = [
    ("Bubble Sort",    bubble_sort),
    ("Selection Sort", selection_sort),
    ("Insertion Sort", insertion_sort),
    ("Merge Sort",     merge_sort),
    ("Quick Sort",     quick_sort),
    ("Counting Sort",  counting_sort),
]

for name, func in algorithms:
    result = func(arr)
    print(f"{name:16} → {result}")

# Bucket sort with integers
print(f"{'Bucket Sort':16} → {bucket_sort(arr)}")

# Python built-in
print(f"{'Python sort()':16} → {sorted(arr)}")
````

## 🏆 Key Takeaways

### When to use which?
```
Small array (n < 10)     → Insertion Sort   (simple & fast for small!)
General purpose          → Merge Sort       (always O(n log n)!)
General purpose (memory) → Quick Sort       (O(log n) space!)
Almost sorted            → Insertion Sort   (O(n) best case!)
Integer range known      → Counting Sort    (O(n+k) super fast!)
Uniform distribution     → Bucket Sort      (O(n+k) average!)
Python built-in          → sorted() / .sort() (TimSort = O(n log n)!)
```

### Simple vs Advanced:
```
Simple (O(n²)):        Bubble, Selection, Insertion
Advanced (O(n log n)): Merge, Quick
Special (O(n+k)):      Counting, Bucket
```

### Real World:
- Python's `sorted()` uses **TimSort** (hybrid Merge + Insertion)
- Java uses **TimSort** for objects, **Dual-Pivot QuickSort** for primitives
- Databases use **external merge sort** for huge datasets
